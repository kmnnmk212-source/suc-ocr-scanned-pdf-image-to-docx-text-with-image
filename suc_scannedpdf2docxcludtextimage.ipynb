````python
#!/usr/bin/env python3
"""
PDF → DOCX Converter مع الحفاظ على التخطيط المكاني
=====================================================
يقوم البرنامج بـ:
1. تحليل كل صفحة PDF وتحديد العناصر (نصوص، صور، جداول، رسومات)
2. رسم مستطيلات حول كل عنصر في صورة معاينة
3. استخراج النصوص كنصوص قابلة للتحرير
4. استخراج الصور والرسوم البيانية كصور
5. الحفاظ على العلاقات المكانية النسبية في ملف DOCX
"""

import sys
import os
import io
import json
import argparse
from pathlib import Path
from dataclasses import dataclass, field
from typing import List, Tuple, Optional
from enum import Enum

import fitz  # PyMuPDF
import pdfplumber
from PIL import Image, ImageDraw, ImageFont
from docx import Document
from docx.shared import Pt, Inches, Emu, RGBColor, Cm
from docx.enum.text import WD_ALIGN_PARAGRAPH
from docx.oxml.ns import qn
from docx.oxml import OxmlElement
import numpy as np

try:
    import pytesseract
    OCR_AVAILABLE = True
except ImportError:
    OCR_AVAILABLE = False

try:
    import cv2
    CV2_AVAILABLE = True
except ImportError:
    CV2_AVAILABLE = False


# ── أنواع العناصر ──────────────────────────────────────────────────────────────

class ElementType(Enum):
    TEXT  = "text"
    IMAGE = "image"
    TABLE = "table"
    CHART = "chart"   # رسم بياني / شكل معقد


# ألوان مستطيلات التمييز لكل نوع (R, G, B)
ELEMENT_COLORS = {
    ElementType.TEXT:  (0,   120, 215),   # أزرق
    ElementType.IMAGE: (0,   180,  90),   # أخضر
    ElementType.TABLE: (220,  80,  30),   # برتقالي
    ElementType.CHART: (160,  32, 240),   # بنفسجي
}

ELEMENT_LABELS = {
    ElementType.TEXT:  "TEXT  نص",
    ElementType.IMAGE: "IMAGE صورة",
    ElementType.TABLE: "TABLE جدول",
    ElementType.CHART: "CHART رسم بياني",
}


# ── هيكل بيانات العنصر ─────────────────────────────────────────────────────────

@dataclass
class PageElement:
    etype:   ElementType
    bbox:    Tuple[float, float, float, float]   # x0, y0, x1, y1  (PDF units)
    content: object = None     # نص str / صورة bytes / بيانات جدول list
    font_info: dict = field(default_factory=dict)


# ── استخراج العناصر من صفحة ────────────────────────────────────────────────────

def is_scanned_page(fitz_page: fitz.Page) -> bool:
    """يحدد إن كانت الصفحة مجرد صورة ممسوحة (سكانر) بدون نص حقيقي."""
    text = fitz_page.get_text("text").strip()
    images = fitz_page.get_images(full=True)
    if len(text) < 15 and len(images) >= 1:
        # تحقق أن الصورة تغطي معظم الصفحة
        page_area = fitz_page.rect.width * fitz_page.rect.height
        for img_info in images:
            xref = img_info[0]
            rects = fitz_page.get_image_rects(xref)
            for r in rects:
                img_area = (r.x1 - r.x0) * (r.y1 - r.y0)
                if img_area / page_area > 0.6:
                    return True
    return len(text) < 15 and len(images) == 0  # صفحة فارغة تماماً من النص أيضاً تحتاج تحقق


def detect_picture_regions(pil_img: Image.Image, scale: float) -> List[Tuple[float, float, float, float]]:
    """
    يكتشف مناطق الصور/الرسومات داخل صفحة سكانر (وليس النص) عبر تحليل
    كثافة الحواف والتباين المحلي. يرجّع bboxes بوحدات PDF (pt).
    """
    if not CV2_AVAILABLE:
        return []

    gray = cv2.cvtColor(np.array(pil_img), cv2.COLOR_RGB2GRAY)
    h, w = gray.shape

    # حواف قوية (كاني) لتمييز الرسومات كثيفة التفاصيل عن سطور النص الرفيعة
    edges = cv2.Canny(gray, 50, 150)

    # تكثيف بسيط لربط تفاصيل الرسم الداخلية فقط (نواة صغيرة لتجنّب الالتحام بنص الكابشن)
    kernel = cv2.getStructuringElement(cv2.MORPH_RECT, (7, 7))
    dilated = cv2.dilate(edges, kernel, iterations=2)

    contours, _ = cv2.findContours(dilated, cv2.RETR_EXTERNAL, cv2.CHAIN_APPROX_SIMPLE)

    regions = []
    page_area = w * h
    for cnt in contours:
        x, y, cw, ch = cv2.boundingRect(cnt)
        area = cw * ch
        area_ratio = area / page_area
        if area_ratio < 0.03 or area_ratio > 0.85:
            continue
        aspect = cw / max(ch, 1)
        if aspect > 4.5 or aspect < 0.2:
            continue
        roi_edges = edges[y:y+ch, x:x+cw]
        edge_density = np.count_nonzero(roi_edges) / max(area, 1)
        if edge_density < 0.06:
            continue

        # تقليم الحافة السفلية بدقة: نتحرك من الأسفل لأعلى ونوقف عند أول صف
        # كثيف الحواف بعد عبور صف فارغ كامل (هامش حقيقي بين الرسم والكابشن)
        row_density = (edges[y:y+ch, x:x+cw] > 0).sum(axis=1) / max(cw, 1)
        new_ch = ch
        i = ch - 1
        seen_content = False
        gap_count = 0
        while i >= 0:
            if row_density[i] > 0.02:
                seen_content = True
                gap_count = 0
            elif seen_content:
                gap_count += 1
                if gap_count > max(3, int(0.01 * h)):
                    new_ch = i
                    break
            i -= 1
        ch = max(new_ch, int((ch) * 0.3))  # حد أدنى أماني

        regions.append((x/scale, y/scale, (x+cw)/scale, (y+ch)/scale))

    # دمج المناطق المتداخلة/المتقاربة
    merged = []
    for r in sorted(regions, key=lambda r: -(r[2]-r[0])*(r[3]-r[1])):
        keep = True
        for m in merged:
            ix0, iy0 = max(r[0], m[0]), max(r[1], m[1])
            ix1, iy1 = min(r[2], m[2]), min(r[3], m[3])
            if ix1 > ix0 and iy1 > iy0:
                inter = (ix1-ix0)*(iy1-iy0)
                if inter / min((r[2]-r[0])*(r[3]-r[1]), (m[2]-m[0])*(m[3]-m[1])) > 0.4:
                    keep = False
                    break
        if keep:
            merged.append(r)
    return merged


def ocr_extract_elements(fitz_page: fitz.Page, lang: str = "ara+eng",
                          scale: float = 3.0) -> List["PageElement"]:
    """يستخرج عناصر نصية وصورية من صفحة ممسوحة (سكانر) باستخدام OCR + كشف بصري."""
    mat = fitz.Matrix(scale, scale)
    pix = fitz_page.get_pixmap(matrix=mat, alpha=False)
    img = Image.frombytes("RGB", [pix.width, pix.height], pix.samples)

    elements: List[PageElement] = []

    # ── 1. كشف مناطق الصور/الرسومات بصرياً ─────────────────────────────────
    picture_regions = detect_picture_regions(img, scale)
    for (px0, py0, px1, py1) in picture_regions:
        cx0, cy0 = int(px0*scale), int(py0*scale)
        cx1, cy1 = int(px1*scale), int(py1*scale)
        if cx1 - cx0 < 5 or cy1 - cy0 < 5:
            print(f"      [تحذير] منطقة صورة مكتشفة بأبعاد غير صالحة ({cx1-cx0}×{cy1-cy0}) — تم تجاهلها")
            continue
        try:
            crop = img.crop((cx0, cy0, cx1, cy1))
            buf = io.BytesIO()
            crop.save(buf, format="PNG")
            img_data = buf.getvalue()
            if len(img_data) < 100:
                print(f"      [تحذير] بيانات صورة صغيرة جداً ({len(img_data)} بايت) — تم تجاهلها")
                continue
            elements.append(PageElement(ElementType.IMAGE, (px0, py0, px1, py1), img_data))
        except Exception as e:
            print(f"      [خطأ] فشل قص/حفظ منطقة صورة: {e}")

    if not OCR_AVAILABLE:
        print("  [تحذير] pytesseract غير متاح؛ لن يتم استخراج النص من الصفحات الممسوحة")
        print("          ثبّت: pip install pytesseract --break-system-packages && apt-get install tesseract-ocr tesseract-ocr-ara")
        elements.sort(key=lambda e: (e.bbox[1], e.bbox[0]))
        return elements

    try:
        ocr_data = pytesseract.image_to_data(
            img, lang=lang, output_type=pytesseract.Output.DICT,
            config="--psm 3"
        )
    except Exception as e:
        print(f"  [خطأ OCR] {e}")
        elements.sort(key=lambda e: (e.bbox[1], e.bbox[0]))
        return elements

    def _inside_picture(x0, y0, x1, y1) -> bool:
        cx, cy = (x0+x1)/2, (y0+y1)/2
        for (px0, py0, px1, py1) in picture_regions:
            if px0 <= cx <= px1 and py0 <= cy <= py1:
                return True
        return False

    n = len(ocr_data["text"])
    # تجميع الكلمات حسب رقم الكتلة + الفقرة + السطر لإعادة بناء أسطر كاملة
    lines = {}
    for i in range(n):
        word = ocr_data["text"][i].strip()
        conf_raw = str(ocr_data.get("conf", ["-1"]*n)[i])
        conf = int(conf_raw) if conf_raw.lstrip("-").isdigit() else -1
        if not word or conf < 25:
            continue
        x, y, w, h = (ocr_data["left"][i], ocr_data["top"][i],
                      ocr_data["width"][i], ocr_data["height"][i])
        # استبعاد الكلمات الواقعة داخل منطقة صورة مكتشفة (غالباً ضجيج OCR على الرسم)
        if _inside_picture(x/scale, y/scale, (x+w)/scale, (y+h)/scale):
            continue
        key = (ocr_data["block_num"][i], ocr_data["par_num"][i], ocr_data["line_num"][i])
        if key not in lines:
            lines[key] = {"words": [], "x0": x, "y0": y, "x1": x+w, "y1": y+h}
        ld = lines[key]
        ld["words"].append(word)
        ld["x0"] = min(ld["x0"], x)
        ld["y0"] = min(ld["y0"], y)
        ld["x1"] = max(ld["x1"], x + w)
        ld["y1"] = max(ld["y1"], y + h)

    text_elements = []
    for key in sorted(lines.keys()):
        ld = lines[key]
        text = " ".join(ld["words"])
        # تحويل من بكسلات الصورة (scale×) إلى وحدات PDF (pt)
        bbox = (ld["x0"]/scale, ld["y0"]/scale, ld["x1"]/scale, ld["y1"]/scale)
        avg_h_pt = (ld["y1"] - ld["y0"]) / scale
        font_size = max(7, min(28, avg_h_pt * 0.8))
        text_elements.append(PageElement(ElementType.TEXT, bbox, text,
                                    {"avg_size": font_size, "ocr": True}))

    text_elements.sort(key=lambda e: (e.bbox[1], e.bbox[0]))

    # دمج الأسطر المتتالية القريبة عمودياً ومتشابهة بحجم الخط في فقرة واحدة
    merged: List[PageElement] = []
    for elem in text_elements:
        if merged:
            prev = merged[-1]
            same_size = abs(prev.font_info.get("avg_size", 12) - elem.font_info.get("avg_size", 12)) < 2
            vertical_gap = elem.bbox[1] - prev.bbox[3]
            line_h = prev.bbox[3] - prev.bbox[1]
            close_enough = 0 <= vertical_gap < line_h * 0.9
            similar_left = abs(elem.bbox[0] - prev.bbox[0]) < 30
            if same_size and close_enough and similar_left:
                prev.content = prev.content.rstrip() + " " + elem.content
                prev.bbox = (min(prev.bbox[0], elem.bbox[0]), prev.bbox[1],
                            max(prev.bbox[2], elem.bbox[2]), elem.bbox[3])
                continue
        merged.append(elem)

    all_elements = elements + merged
    all_elements.sort(key=lambda e: (round(e.bbox[1] / 8), e.bbox[0]))
    return all_elements


def extract_elements(fitz_page: fitz.Page, plumber_page) -> List[PageElement]:
    """يستخرج جميع العناصر من الصفحة، مع تفرّع آلي لـ OCR إن كانت الصفحة ممسوحة."""
    if is_scanned_page(fitz_page):
        print("      🔍 صفحة ممسوحة (سكانر) — تشغيل OCR + كشف الصور...")
        return ocr_extract_elements(fitz_page)
    return _extract_elements_native(fitz_page, plumber_page)


def _extract_elements_native(fitz_page: fitz.Page, plumber_page) -> List[PageElement]:
    """يستخرج جميع العناصر من الصفحة ويصنّفها."""
    elements: List[PageElement] = []
    used_rects = []   # مناطق مستخدمة لتجنب التداخل

    # ── 1. الجداول (pdfplumber أفضل للجداول) ──────────────────────────────
    try:
        tables = plumber_page.extract_tables(
            {"vertical_strategy": "lines", "horizontal_strategy": "lines"}
        )
        table_bboxes = []
        for t_obj in plumber_page.find_tables():
            bb = t_obj.bbox   # (x0,top,x1,bottom)  pdfplumber coords
            # تحويل إحداثيات pdfplumber → fitz  (y محور مقلوب)
            page_h = float(plumber_page.height)
            fx0, fy0 = bb[0], page_h - bb[3]
            fx1, fy1 = bb[2], page_h - bb[1]
            table_bboxes.append((fx0, fy0, fx1, fy1))
            data = t_obj.extract()
            elements.append(PageElement(ElementType.TABLE, (fx0,fy0,fx1,fy1), data))
            used_rects.append(fitz.Rect(fx0, fy0, fx1, fy1))
    except Exception as e:
        print(f"  [تحذير] استخراج الجداول: {e}")

    # ── 2. الصور المضمّنة ────────────────────────────────────────────────────
    img_list = fitz_page.get_images(full=True)
    for img_info in img_list:
        xref = img_info[0]
        # الحصول على bbox الصورة داخل الصفحة
        img_rects = fitz_page.get_image_rects(xref)
        if not img_rects:
            continue
        r = img_rects[0]
        # تجاهل صور الخلفية الكبيرة جداً (>85% من الصفحة)
        page_area = fitz_page.rect.width * fitz_page.rect.height
        img_area  = (r.x1 - r.x0) * (r.y1 - r.y0)
        if img_area / page_area > 0.85:
            continue
        # استخراج بيانات الصورة
        try:
            base_img = fitz_page.parent.extract_image(xref)
            img_bytes = base_img["image"]
            bbox = (r.x0, r.y0, r.x1, r.y1)

            # هل هي رسم بياني؟ نفحص بالحجم والموقع
            etype = ElementType.IMAGE
            # رسوم بيانية عادةً: عرض > 200pt وارتفاع > 150pt
            if (r.x1-r.x0) > 200 and (r.y1-r.y0) > 150:
                etype = ElementType.CHART

            elements.append(PageElement(etype, bbox, img_bytes))
            used_rects.append(r)
        except Exception:
            pass

    # ── 3. كتل النصوص ───────────────────────────────────────────────────────
    blocks = fitz_page.get_text("dict", flags=fitz.TEXT_PRESERVE_WHITESPACE)["blocks"]
    for blk in blocks:
        if blk["type"] != 0:   # 0 = نص
            continue
        r = fitz.Rect(blk["bbox"])
        # تحقق من أن الكتلة لا تتداخل مع جدول/صورة موجودة
        overlap = any(r.intersects(ur) and _overlap_ratio(r, ur) > 0.5 for ur in used_rects)
        if overlap:
            continue

        # تجميع معلومات الخط
        lines_text = []
        font_sizes = []
        for line in blk.get("lines", []):
            for span in line.get("spans", []):
                lines_text.append(span["text"])
                font_sizes.append(span["size"])

        full_text = "\n".join(lines_text).strip()
        if not full_text:
            continue

        avg_size = sum(font_sizes) / len(font_sizes) if font_sizes else 12
        bbox = (r.x0, r.y0, r.x1, r.y1)
        elem = PageElement(ElementType.TEXT, bbox, full_text,
                           {"avg_size": avg_size, "spans": blk})
        elements.append(elem)

    # ── ترتيب العناصر من أعلى لأسفل ثم يسار لأيمن ─────────────────────────
    elements.sort(key=lambda e: (round(e.bbox[1] / 10), e.bbox[0]))
    return elements


def _overlap_ratio(r1: fitz.Rect, r2: fitz.Rect) -> float:
    inter = r1 & r2
    if inter.is_empty:
        return 0.0
    return inter.get_area() / min(r1.get_area(), r2.get_area() + 1e-9)


# ── رسم مستطيلات التمييز على صورة معاينة ────────────────────────────────────

def annotate_page_image(fitz_page: fitz.Page, elements: List[PageElement],
                        scale: float = 2.0) -> Image.Image:
    """يرسم صورة للصفحة مع مستطيلات ترميزية حول كل عنصر."""
    mat = fitz.Matrix(scale, scale)
    pix = fitz_page.get_pixmap(matrix=mat, alpha=False)
    img = Image.frombytes("RGB", [pix.width, pix.height], pix.samples)
    draw = ImageDraw.Draw(img, "RGBA")

    try:
        font = ImageFont.truetype("/usr/share/fonts/truetype/dejavu/DejaVuSans-Bold.ttf", 14)
    except Exception:
        font = ImageFont.load_default()

    for elem in elements:
        x0, y0, x1, y1 = [c * scale for c in elem.bbox]
        r, g, b = ELEMENT_COLORS[elem.etype]
        # مستطيل شبه شفاف
        draw.rectangle([x0, y0, x1, y1],
                       outline=(r, g, b, 255),
                       fill=(r, g, b, 25),
                       width=2)
        # تسمية في أعلى اليسار
        label = ELEMENT_LABELS[elem.etype]
        tx, ty = x0 + 3, y0 + 2
        # خلفية للنص
        draw.rectangle([tx - 1, ty - 1, tx + len(label)*7 + 2, ty + 16],
                       fill=(r, g, b, 180))
        draw.text((tx, ty), label, fill="white", font=font)

    return img


# ── إنشاء ملف DOCX مع الحفاظ على التخطيط ────────────────────────────────────

def _set_cell_border(cell, **kwargs):
    """ضبط حدود خلية جدول."""
    tc = cell._tc
    tcPr = tc.get_or_add_tcPr()
    tcBorders = OxmlElement("w:tcBorders")
    for edge in ("top", "left", "bottom", "right", "insideH", "insideV"):
        edge_data = kwargs.get(edge)
        if edge_data:
            tag = OxmlElement(f"w:{edge}")
            for attr, val in edge_data.items():
                tag.set(qn(f"w:{attr}"), str(val))
            tcBorders.append(tag)
    tcPr.append(tcBorders)


def _contains_arabic(text: str) -> bool:
    return any('\u0600' <= ch <= '\u06FF' for ch in text)


def _set_rtl(paragraph):
    """يضبط اتجاه الفقرة من اليمين لليسار (للنصوص العربية)."""
    pPr = paragraph._p.get_or_add_pPr()
    bidi = OxmlElement("w:bidi")
    pPr.append(bidi)


def add_element_to_doc(doc: Document, elem: PageElement,
                        page_w_pt: float, page_h_pt: float,
                        content_w_in: float):
    """يضيف عنصراً واحداً إلى مستند الـ DOCX مع الحفاظ على نسبه المكانية."""

    x0, y0, x1, y1 = elem.bbox
    # نسبة العرض
    rel_x   = x0 / page_w_pt          # من أين يبدأ (0→1)
    rel_w   = (x1 - x0) / page_w_pt   # عرضه النسبي
    elem_w_in = content_w_in * rel_w

    # ── نص ──────────────────────────────────────────────────────────────────
    if elem.etype == ElementType.TEXT:
        para = doc.add_paragraph()
        run  = para.add_run(elem.content)
        avg_sz = elem.font_info.get("avg_size", 12)
        run.font.size = Pt(max(6, avg_sz))

        is_arabic = _contains_arabic(elem.content)
        if is_arabic:
            run.font.name = "Arial"
            _set_rtl(para)
            # محاذاة معكوسة للعربية: يمين = بداية النص
            if rel_x > 0.55:
                para.alignment = WD_ALIGN_PARAGRAPH.LEFT
            elif rel_x > 0.25:
                para.alignment = WD_ALIGN_PARAGRAPH.CENTER
            else:
                para.alignment = WD_ALIGN_PARAGRAPH.RIGHT
        else:
            if rel_x > 0.55:
                para.alignment = WD_ALIGN_PARAGRAPH.RIGHT
            elif rel_x > 0.25:
                para.alignment = WD_ALIGN_PARAGRAPH.CENTER
            else:
                para.alignment = WD_ALIGN_PARAGRAPH.LEFT
        # مسافة بادئة بسيطة
        left_indent = Inches(content_w_in * rel_x * 0.7)
        para.paragraph_format.left_indent = min(left_indent, Inches(content_w_in * 0.5))

    # ── صورة / رسم بياني ─────────────────────────────────────────────────────
    elif elem.etype in (ElementType.IMAGE, ElementType.CHART):
        try:
            img_bytes = elem.content
            if not img_bytes:
                print(f"      [تحذير] عنصر صورة بدون بيانات — تم تجاهله")
                return
            img_stream = io.BytesIO(img_bytes)
            # تحقق من صحة الصورة بدون استخدام verify() الذي يُفسد الـ stream في بعض إصدارات Pillow
            pil_img = Image.open(img_stream)
            pil_img.load()          # يقرأ البيانات فعلياً للتأكد من سلامتها
            img_stream.seek(0)      # نعيد المؤشر لأول الستريم بعد القراءة

            para = doc.add_paragraph()
            if rel_x > 0.55:
                para.alignment = WD_ALIGN_PARAGRAPH.RIGHT
            elif rel_x > 0.25:
                para.alignment = WD_ALIGN_PARAGRAPH.CENTER
            else:
                para.alignment = WD_ALIGN_PARAGRAPH.LEFT

            run = para.add_run()
            display_w = min(elem_w_in, content_w_in)
            if display_w <= 0:
                display_w = content_w_in * 0.5
            run.add_picture(img_stream, width=Inches(display_w))
        except Exception as e:
            import traceback
            print(f"      [خطأ] فشل إضافة صورة ({elem.etype.value}): {e}")
            traceback.print_exc()
            para = doc.add_paragraph(f"[فشل استخراج صورة: {e}]")

    # ── جدول ─────────────────────────────────────────────────────────────────
    elif elem.etype == ElementType.TABLE:
        data = elem.content
        if not data:
            return
        # تنظيف البيانات
        rows = [[str(c or "") for c in row] for row in data]
        if not rows or not rows[0]:
            return

        ncols = max(len(r) for r in rows)
        nrows = len(rows)

        tbl = doc.add_table(rows=nrows, cols=ncols)
        tbl.style = "Table Grid"

        for ri, row in enumerate(rows):
            for ci, cell_txt in enumerate(row):
                if ci >= ncols:
                    break
                cell = tbl.cell(ri, ci)
                cell.text = cell_txt
                # تمييز الصف الأول (رأس الجدول)
                if ri == 0:
                    for run in cell.paragraphs[0].runs:
                        run.bold = True

        doc.add_paragraph()   # سطر فراغ بعد الجدول


# ── الدالة الرئيسية ───────────────────────────────────────────────────────────

def convert_pdf_to_docx(input_pdf: str, output_docx: str,
                         save_preview: bool = True,
                         preview_dir: str = None):
    """
    يحوّل ملف PDF إلى DOCX مع الحفاظ على التخطيط المكاني.
    """
    input_pdf  = str(input_pdf)
    output_docx = str(output_docx)

    print(f"\n{'='*60}")
    print(f"  📄 تحليل: {input_pdf}")
    print(f"{'='*60}")

    fitz_doc   = fitz.open(input_pdf)
    doc        = Document()

    # ── إعدادات الصفحة (A4 افتراضي) ────────────────────────────────────────
    section = doc.sections[0]
    section.page_width  = Cm(21)
    section.page_height = Cm(29.7)
    margin = Cm(2.0)
    section.top_margin    = margin
    section.bottom_margin = margin
    section.left_margin   = margin
    section.right_margin  = margin
    content_w_in = (21 - 4) / 2.54   # عرض المحتوى بالإنش

    preview_paths = []

    with pdfplumber.open(input_pdf) as plumber_doc:
        for page_num in range(len(fitz_doc)):
            fitz_page    = fitz_doc[page_num]
            plumber_page = plumber_doc.pages[page_num]

            pw = float(fitz_page.rect.width)
            ph = float(fitz_page.rect.height)

            print(f"\n  ── صفحة {page_num+1} ({pw:.0f}×{ph:.0f} pt) ──")

            # استخراج العناصر
            elements = extract_elements(fitz_page, plumber_page)

            counts = {}
            for e in elements:
                counts[e.etype.name] = counts.get(e.etype.name, 0) + 1
            for k, v in counts.items():
                print(f"      {k}: {v}")

            # صورة المعاينة مع المستطيلات
            if save_preview:
                ann_img = annotate_page_image(fitz_page, elements, scale=2.0)
                if preview_dir:
                    os.makedirs(preview_dir, exist_ok=True)
                    p_path = os.path.join(preview_dir, f"page_{page_num+1:03d}_preview.png")
                else:
                    p_path = os.path.splitext(output_docx)[0] + f"_page{page_num+1}_preview.png"
                ann_img.save(p_path, "PNG")
                preview_paths.append(p_path)
                print(f"      💾 معاينة: {p_path}")

            # إضافة عنوان الصفحة
            if page_num > 0:
                doc.add_page_break()
            heading = doc.add_paragraph(f"صفحة {page_num+1}")
            heading.style = "Heading 2"

            # إضافة العناصر إلى DOCX
            for elem in elements:
                add_element_to_doc(doc, elem, pw, ph, content_w_in)

    # حفظ DOCX
    doc.save(output_docx)
    print(f"\n✅ تم الحفظ: {output_docx}")
    if preview_paths:
        print(f"✅ صور المعاينة: {len(preview_paths)} ملف")

    fitz_doc.close()
    return output_docx, preview_paths


# ── نقطة الدخول ──────────────────────────────────────────────────────────────

def main():
    parser = argparse.ArgumentParser(
        description="تحويل PDF إلى DOCX مع الحفاظ على التخطيط المكاني"
    )
    parser.add_argument("input",  help="ملف PDF المدخل")
    parser.add_argument("output", nargs="?", help="ملف DOCX المخرج (اختياري)")
    parser.add_argument("--no-preview", action="store_true",
                        help="إيقاف حفظ صور المعاينة")
    parser.add_argument("--preview-dir", default=None,
                        help="مجلد لحفظ صور المعاينة")
    args = parser.parse_args()

    inp  = Path(args.input)
    out  = Path(args.output) if args.output else inp.with_suffix(".docx")

    convert_pdf_to_docx(
        str(inp), str(out),
        save_preview  = not args.no_preview,
        preview_dir   = args.preview_dir,
    )


if __name__ == "__main__":
    main()
````

````python
aa.py








#!/usr/bin/env python3
"""
PDF → DOCX Converter مع الحفاظ على التخطيط المكاني
=====================================================
يقوم البرنامج بـ:
1. تحليل كل صفحة PDF وتحديد العناصر (نصوص، صور، جداول، رسومات)
2. رسم مستطيلات حول كل عنصر في صورة معاينة
3. استخراج النصوص كنصوص قابلة للتحرير
4. استخراج الصور والرسوم البيانية كصور
5. الحفاظ على العلاقات المكانية النسبية في ملف DOCX
"""

import sys
import os
import io
import json
import argparse
from pathlib import Path
from dataclasses import dataclass, field
from typing import List, Tuple, Optional
from enum import Enum

import fitz  # PyMuPDF
import pdfplumber
from PIL import Image, ImageDraw, ImageFont
from docx import Document
from docx.shared import Pt, Inches, Emu, RGBColor, Cm
from docx.enum.text import WD_ALIGN_PARAGRAPH
from docx.oxml.ns import qn
from docx.oxml import OxmlElement
import numpy as np

try:
    import pytesseract
    OCR_AVAILABLE = True
except ImportError:
    OCR_AVAILABLE = False

try:
    import cv2
    CV2_AVAILABLE = True
except ImportError:
    CV2_AVAILABLE = False


# ── أنواع العناصر ──────────────────────────────────────────────────────────────

class ElementType(Enum):
    TEXT  = "text"
    IMAGE = "image"
    TABLE = "table"
    CHART = "chart"   # رسم بياني / شكل معقد


# ألوان مستطيلات التمييز لكل نوع (R, G, B)
ELEMENT_COLORS = {
    ElementType.TEXT:  (0,   120, 215),   # أزرق
    ElementType.IMAGE: (0,   180,  90),   # أخضر
    ElementType.TABLE: (220,  80,  30),   # برتقالي
    ElementType.CHART: (160,  32, 240),   # بنفسجي
}

ELEMENT_LABELS = {
    ElementType.TEXT:  "TEXT  نص",
    ElementType.IMAGE: "IMAGE صورة",
    ElementType.TABLE: "TABLE جدول",
    ElementType.CHART: "CHART رسم بياني",
}


# ── هيكل بيانات العنصر ─────────────────────────────────────────────────────────

@dataclass
class PageElement:
    etype:   ElementType
    bbox:    Tuple[float, float, float, float]   # x0, y0, x1, y1  (PDF units)
    content: object = None     # نص str / صورة bytes / بيانات جدول list
    font_info: dict = field(default_factory=dict)


# ── استخراج العناصر من صفحة ────────────────────────────────────────────────────

def is_scanned_page(fitz_page: fitz.Page) -> bool:
    """يحدد إن كانت الصفحة مجرد صورة ممسوحة (سكانر) بدون نص حقيقي."""
    text = fitz_page.get_text("text").strip()
    images = fitz_page.get_images(full=True)
    if len(text) < 15 and len(images) >= 1:
        # تحقق أن الصورة تغطي معظم الصفحة
        page_area = fitz_page.rect.width * fitz_page.rect.height
        for img_info in images:
            xref = img_info[0]
            rects = fitz_page.get_image_rects(xref)
            for r in rects:
                img_area = (r.x1 - r.x0) * (r.y1 - r.y0)
                if img_area / page_area > 0.6:
                    return True
    return len(text) < 15 and len(images) == 0  # صفحة فارغة تماماً من النص أيضاً تحتاج تحقق


def detect_picture_regions(pil_img: Image.Image, scale: float) -> List[Tuple[float, float, float, float]]:
    """
    يكتشف مناطق الصور/الرسومات داخل صفحة سكانر (وليس النص) عبر تحليل
    كثافة الحواف والتباين المحلي. يرجّع bboxes بوحدات PDF (pt).
    """
    if not CV2_AVAILABLE:
        return []

    gray = cv2.cvtColor(np.array(pil_img), cv2.COLOR_RGB2GRAY)
    h, w = gray.shape

    # حواف قوية (كاني) لتمييز الرسومات كثيفة التفاصيل عن سطور النص الرفيعة
    edges = cv2.Canny(gray, 50, 150)

    # تكثيف بسيط لربط تفاصيل الرسم الداخلية فقط (نواة صغيرة لتجنّب الالتحام بنص الكابشن)
    kernel = cv2.getStructuringElement(cv2.MORPH_RECT, (7, 7))
    dilated = cv2.dilate(edges, kernel, iterations=2)

    contours, _ = cv2.findContours(dilated, cv2.RETR_EXTERNAL, cv2.CHAIN_APPROX_SIMPLE)

    regions = []
    page_area = w * h
    for cnt in contours:
        x, y, cw, ch = cv2.boundingRect(cnt)
        area = cw * ch
        area_ratio = area / page_area
        if area_ratio < 0.03 or area_ratio > 0.85:
            continue
        aspect = cw / max(ch, 1)
        if aspect > 4.5 or aspect < 0.2:
            continue
        roi_edges = edges[y:y+ch, x:x+cw]
        edge_density = np.count_nonzero(roi_edges) / max(area, 1)
        if edge_density < 0.06:
            continue

        # تقليم الحافة السفلية بدقة: نتحرك من الأسفل لأعلى ونوقف عند أول صف
        # كثيف الحواف بعد عبور صف فارغ كامل (هامش حقيقي بين الرسم والكابشن)
        row_density = (edges[y:y+ch, x:x+cw] > 0).sum(axis=1) / max(cw, 1)
        new_ch = ch
        i = ch - 1
        seen_content = False
        gap_count = 0
        while i >= 0:
            if row_density[i] > 0.02:
                seen_content = True
                gap_count = 0
            elif seen_content:
                gap_count += 1
                if gap_count > max(3, int(0.01 * h)):
                    new_ch = i
                    break
            i -= 1
        ch = max(new_ch, int((ch) * 0.3))  # حد أدنى أماني

        regions.append((x/scale, y/scale, (x+cw)/scale, (y+ch)/scale))

    # دمج المناطق المتداخلة/المتقاربة
    merged = []
    for r in sorted(regions, key=lambda r: -(r[2]-r[0])*(r[3]-r[1])):
        keep = True
        for m in merged:
            ix0, iy0 = max(r[0], m[0]), max(r[1], m[1])
            ix1, iy1 = min(r[2], m[2]), min(r[3], m[3])
            if ix1 > ix0 and iy1 > iy0:
                inter = (ix1-ix0)*(iy1-iy0)
                if inter / min((r[2]-r[0])*(r[3]-r[1]), (m[2]-m[0])*(m[3]-m[1])) > 0.4:
                    keep = False
                    break
        if keep:
            merged.append(r)
    return merged


def ocr_extract_elements(fitz_page: fitz.Page, lang: str = "ara+eng",
                          scale: float = 3.0) -> List["PageElement"]:
    """يستخرج عناصر نصية وصورية من صفحة ممسوحة (سكانر) باستخدام OCR + كشف بصري."""
    mat = fitz.Matrix(scale, scale)
    pix = fitz_page.get_pixmap(matrix=mat, alpha=False)
    img = Image.frombytes("RGB", [pix.width, pix.height], pix.samples)

    elements: List[PageElement] = []

    # ── 1. كشف مناطق الصور/الرسومات بصرياً ─────────────────────────────────
    picture_regions = detect_picture_regions(img, scale)
    for (px0, py0, px1, py1) in picture_regions:
        cx0, cy0 = int(px0*scale), int(py0*scale)
        cx1, cy1 = int(px1*scale), int(py1*scale)
        if cx1 - cx0 < 5 or cy1 - cy0 < 5:
            print(f"      [تحذير] منطقة صورة مكتشفة بأبعاد غير صالحة ({cx1-cx0}×{cy1-cy0}) — تم تجاهلها")
            continue
        try:
            crop = img.crop((cx0, cy0, cx1, cy1))
            buf = io.BytesIO()
            crop.save(buf, format="PNG")
            img_data = buf.getvalue()
            if len(img_data) < 100:
                print(f"      [تحذير] بيانات صورة صغيرة جداً ({len(img_data)} بايت) — تم تجاهلها")
                continue
            elements.append(PageElement(ElementType.IMAGE, (px0, py0, px1, py1), img_data))
        except Exception as e:
            print(f"      [خطأ] فشل قص/حفظ منطقة صورة: {e}")

    if not OCR_AVAILABLE:
        print("  [تحذير] pytesseract غير متاح؛ لن يتم استخراج النص من الصفحات الممسوحة")
        print("          ثبّت: pip install pytesseract --break-system-packages && apt-get install tesseract-ocr tesseract-ocr-ara")
        elements.sort(key=lambda e: (e.bbox[1], e.bbox[0]))
        return elements

    try:
        ocr_data = pytesseract.image_to_data(
            img, lang=lang, output_type=pytesseract.Output.DICT,
            config="--psm 3"
        )
    except Exception as e:
        print(f"  [خطأ OCR] {e}")
        elements.sort(key=lambda e: (e.bbox[1], e.bbox[0]))
        return elements

    def _inside_picture(x0, y0, x1, y1) -> bool:
        cx, cy = (x0+x1)/2, (y0+y1)/2
        for (px0, py0, px1, py1) in picture_regions:
            if px0 <= cx <= px1 and py0 <= cy <= py1:
                return True
        return False

    n = len(ocr_data["text"])
    # تجميع الكلمات حسب رقم الكتلة + الفقرة + السطر لإعادة بناء أسطر كاملة
    lines = {}
    for i in range(n):
        word = ocr_data["text"][i].strip()
        conf_raw = str(ocr_data.get("conf", ["-1"]*n)[i])
        conf = int(conf_raw) if conf_raw.lstrip("-").isdigit() else -1
        if not word or conf < 25:
            continue
        x, y, w, h = (ocr_data["left"][i], ocr_data["top"][i],
                      ocr_data["width"][i], ocr_data["height"][i])
        # استبعاد الكلمات الواقعة داخل منطقة صورة مكتشفة (غالباً ضجيج OCR على الرسم)
        if _inside_picture(x/scale, y/scale, (x+w)/scale, (y+h)/scale):
            continue
        key = (ocr_data["block_num"][i], ocr_data["par_num"][i], ocr_data["line_num"][i])
        if key not in lines:
            lines[key] = {"words": [], "x0": x, "y0": y, "x1": x+w, "y1": y+h}
        ld = lines[key]
        ld["words"].append(word)
        ld["x0"] = min(ld["x0"], x)
        ld["y0"] = min(ld["y0"], y)
        ld["x1"] = max(ld["x1"], x + w)
        ld["y1"] = max(ld["y1"], y + h)

    text_elements = []
    for key in sorted(lines.keys()):
        ld = lines[key]
        text = " ".join(ld["words"])
        # تحويل من بكسلات الصورة (scale×) إلى وحدات PDF (pt)
        bbox = (ld["x0"]/scale, ld["y0"]/scale, ld["x1"]/scale, ld["y1"]/scale)
        avg_h_pt = (ld["y1"] - ld["y0"]) / scale
        font_size = max(7, min(28, avg_h_pt * 0.8))
        text_elements.append(PageElement(ElementType.TEXT, bbox, text,
                                    {"avg_size": font_size, "ocr": True}))

    text_elements.sort(key=lambda e: (e.bbox[1], e.bbox[0]))

    # دمج الأسطر المتتالية القريبة عمودياً ومتشابهة بحجم الخط في فقرة واحدة
    merged: List[PageElement] = []
    for elem in text_elements:
        if merged:
            prev = merged[-1]
            same_size = abs(prev.font_info.get("avg_size", 12) - elem.font_info.get("avg_size", 12)) < 2
            vertical_gap = elem.bbox[1] - prev.bbox[3]
            line_h = prev.bbox[3] - prev.bbox[1]
            close_enough = 0 <= vertical_gap < line_h * 0.9
            similar_left = abs(elem.bbox[0] - prev.bbox[0]) < 30
            if same_size and close_enough and similar_left:
                prev.content = prev.content.rstrip() + " " + elem.content
                prev.bbox = (min(prev.bbox[0], elem.bbox[0]), prev.bbox[1],
                            max(prev.bbox[2], elem.bbox[2]), elem.bbox[3])
                continue
        merged.append(elem)

    all_elements = elements + merged
    all_elements.sort(key=lambda e: (round(e.bbox[1] / 8), e.bbox[0]))
    return all_elements


def extract_elements(fitz_page: fitz.Page, plumber_page) -> List[PageElement]:
    """يستخرج جميع العناصر من الصفحة، مع تفرّع آلي لـ OCR إن كانت الصفحة ممسوحة."""
    if is_scanned_page(fitz_page):
        print("      🔍 صفحة ممسوحة (سكانر) — تشغيل OCR + كشف الصور...")
        return ocr_extract_elements(fitz_page)
    return _extract_elements_native(fitz_page, plumber_page)


def _extract_elements_native(fitz_page: fitz.Page, plumber_page) -> List[PageElement]:
    """يستخرج جميع العناصر من الصفحة ويصنّفها."""
    elements: List[PageElement] = []
    used_rects = []   # مناطق مستخدمة لتجنب التداخل

    # ── 1. الجداول (pdfplumber أفضل للجداول) ──────────────────────────────
    try:
        tables = plumber_page.extract_tables(
            {"vertical_strategy": "lines", "horizontal_strategy": "lines"}
        )
        table_bboxes = []
        for t_obj in plumber_page.find_tables():
            bb = t_obj.bbox   # (x0,top,x1,bottom)  pdfplumber coords
            # تحويل إحداثيات pdfplumber → fitz  (y محور مقلوب)
            page_h = float(plumber_page.height)
            fx0, fy0 = bb[0], page_h - bb[3]
            fx1, fy1 = bb[2], page_h - bb[1]
            table_bboxes.append((fx0, fy0, fx1, fy1))
            data = t_obj.extract()
            elements.append(PageElement(ElementType.TABLE, (fx0,fy0,fx1,fy1), data))
            used_rects.append(fitz.Rect(fx0, fy0, fx1, fy1))
    except Exception as e:
        print(f"  [تحذير] استخراج الجداول: {e}")

    # ── 2. الصور المضمّنة ────────────────────────────────────────────────────
    img_list = fitz_page.get_images(full=True)
    for img_info in img_list:
        xref = img_info[0]
        # الحصول على bbox الصورة داخل الصفحة
        img_rects = fitz_page.get_image_rects(xref)
        if not img_rects:
            continue
        r = img_rects[0]
        # تجاهل صور الخلفية الكبيرة جداً (>85% من الصفحة)
        page_area = fitz_page.rect.width * fitz_page.rect.height
        img_area  = (r.x1 - r.x0) * (r.y1 - r.y0)
        if img_area / page_area > 0.85:
            continue
        # استخراج بيانات الصورة
        try:
            base_img = fitz_page.parent.extract_image(xref)
            img_bytes = base_img["image"]
            bbox = (r.x0, r.y0, r.x1, r.y1)

            # هل هي رسم بياني؟ نفحص بالحجم والموقع
            etype = ElementType.IMAGE
            # رسوم بيانية عادةً: عرض > 200pt وارتفاع > 150pt
            if (r.x1-r.x0) > 200 and (r.y1-r.y0) > 150:
                etype = ElementType.CHART

            elements.append(PageElement(etype, bbox, img_bytes))
            used_rects.append(r)
        except Exception:
            pass

    # ── 3. كتل النصوص ───────────────────────────────────────────────────────
    blocks = fitz_page.get_text("dict", flags=fitz.TEXT_PRESERVE_WHITESPACE)["blocks"]
    for blk in blocks:
        if blk["type"] != 0:   # 0 = نص
            continue
        r = fitz.Rect(blk["bbox"])
        # تحقق من أن الكتلة لا تتداخل مع جدول/صورة موجودة
        overlap = any(r.intersects(ur) and _overlap_ratio(r, ur) > 0.5 for ur in used_rects)
        if overlap:
            continue

        # تجميع معلومات الخط
        lines_text = []
        font_sizes = []
        for line in blk.get("lines", []):
            for span in line.get("spans", []):
                lines_text.append(span["text"])
                font_sizes.append(span["size"])

        full_text = "\n".join(lines_text).strip()
        if not full_text:
            continue

        avg_size = sum(font_sizes) / len(font_sizes) if font_sizes else 12
        bbox = (r.x0, r.y0, r.x1, r.y1)
        elem = PageElement(ElementType.TEXT, bbox, full_text,
                           {"avg_size": avg_size, "spans": blk})
        elements.append(elem)

    # ── ترتيب العناصر من أعلى لأسفل ثم يسار لأيمن ─────────────────────────
    elements.sort(key=lambda e: (round(e.bbox[1] / 10), e.bbox[0]))
    return elements


def _overlap_ratio(r1: fitz.Rect, r2: fitz.Rect) -> float:
    inter = r1 & r2
    if inter.is_empty:
        return 0.0
    return inter.get_area() / min(r1.get_area(), r2.get_area() + 1e-9)


# ── رسم مستطيلات التمييز على صورة معاينة ────────────────────────────────────

def annotate_page_image(fitz_page: fitz.Page, elements: List[PageElement],
                        scale: float = 2.0) -> Image.Image:
    """يرسم صورة للصفحة مع مستطيلات ترميزية حول كل عنصر."""
    mat = fitz.Matrix(scale, scale)
    pix = fitz_page.get_pixmap(matrix=mat, alpha=False)
    img = Image.frombytes("RGB", [pix.width, pix.height], pix.samples)
    draw = ImageDraw.Draw(img, "RGBA")

    try:
        font = ImageFont.truetype("/usr/share/fonts/truetype/dejavu/DejaVuSans-Bold.ttf", 14)
    except Exception:
        font = ImageFont.load_default()

    for elem in elements:
        x0, y0, x1, y1 = [c * scale for c in elem.bbox]
        r, g, b = ELEMENT_COLORS[elem.etype]
        # مستطيل شبه شفاف
        draw.rectangle([x0, y0, x1, y1],
                       outline=(r, g, b, 255),
                       fill=(r, g, b, 25),
                       width=2)
        # تسمية في أعلى اليسار
        label = ELEMENT_LABELS[elem.etype]
        tx, ty = x0 + 3, y0 + 2
        # خلفية للنص
        draw.rectangle([tx - 1, ty - 1, tx + len(label)*7 + 2, ty + 16],
                       fill=(r, g, b, 180))
        draw.text((tx, ty), label, fill="white", font=font)

    return img


# ── إنشاء ملف DOCX مع الحفاظ على التخطيط ────────────────────────────────────

def _set_cell_border(cell, **kwargs):
    """ضبط حدود خلية جدول."""
    tc = cell._tc
    tcPr = tc.get_or_add_tcPr()
    tcBorders = OxmlElement("w:tcBorders")
    for edge in ("top", "left", "bottom", "right", "insideH", "insideV"):
        edge_data = kwargs.get(edge)
        if edge_data:
            tag = OxmlElement(f"w:{edge}")
            for attr, val in edge_data.items():
                tag.set(qn(f"w:{attr}"), str(val))
            tcBorders.append(tag)
    tcPr.append(tcBorders)


def _contains_arabic(text: str) -> bool:
    return any('\u0600' <= ch <= '\u06FF' for ch in text)


def _set_rtl(paragraph):
    """يضبط اتجاه الفقرة من اليمين لليسار (للنصوص العربية)."""
    pPr = paragraph._p.get_or_add_pPr()
    bidi = OxmlElement("w:bidi")
    pPr.append(bidi)


def add_element_to_doc(doc: Document, elem: PageElement,
                        page_w_pt: float, page_h_pt: float,
                        content_w_in: float):
    """يضيف عنصراً واحداً إلى مستند الـ DOCX مع الحفاظ على نسبه المكانية."""

    x0, y0, x1, y1 = elem.bbox
    # نسبة العرض
    rel_x   = x0 / page_w_pt          # من أين يبدأ (0→1)
    rel_w   = (x1 - x0) / page_w_pt   # عرضه النسبي
    elem_w_in = content_w_in * rel_w

    # ── نص ──────────────────────────────────────────────────────────────────
    if elem.etype == ElementType.TEXT:
        para = doc.add_paragraph()
        run  = para.add_run(elem.content)
        avg_sz = elem.font_info.get("avg_size", 12)
        run.font.size = Pt(max(6, avg_sz))

        is_arabic = _contains_arabic(elem.content)
        if is_arabic:
            run.font.name = "Arial"
            _set_rtl(para)
            # محاذاة معكوسة للعربية: يمين = بداية النص
            if rel_x > 0.55:
                para.alignment = WD_ALIGN_PARAGRAPH.LEFT
            elif rel_x > 0.25:
                para.alignment = WD_ALIGN_PARAGRAPH.CENTER
            else:
                para.alignment = WD_ALIGN_PARAGRAPH.RIGHT
        else:
            if rel_x > 0.55:
                para.alignment = WD_ALIGN_PARAGRAPH.RIGHT
            elif rel_x > 0.25:
                para.alignment = WD_ALIGN_PARAGRAPH.CENTER
            else:
                para.alignment = WD_ALIGN_PARAGRAPH.LEFT
        # مسافة بادئة بسيطة
        left_indent = Inches(content_w_in * rel_x * 0.7)
        para.paragraph_format.left_indent = min(left_indent, Inches(content_w_in * 0.5))

    # ── صورة / رسم بياني ─────────────────────────────────────────────────────
    elif elem.etype in (ElementType.IMAGE, ElementType.CHART):
        try:
            img_bytes = elem.content
            if not img_bytes:
                print(f"      [تحذير] عنصر صورة بدون بيانات — تم تجاهله")
                return
            img_stream = io.BytesIO(img_bytes)
            # تحقق من صحة الصورة بدون استخدام verify() الذي يُفسد الـ stream في بعض إصدارات Pillow
            pil_img = Image.open(img_stream)
            pil_img.load()          # يقرأ البيانات فعلياً للتأكد من سلامتها
            img_stream.seek(0)      # نعيد المؤشر لأول الستريم بعد القراءة

            para = doc.add_paragraph()
            if rel_x > 0.55:
                para.alignment = WD_ALIGN_PARAGRAPH.RIGHT
            elif rel_x > 0.25:
                para.alignment = WD_ALIGN_PARAGRAPH.CENTER
            else:
                para.alignment = WD_ALIGN_PARAGRAPH.LEFT

            run = para.add_run()
            display_w = min(elem_w_in, content_w_in)
            if display_w <= 0:
                display_w = content_w_in * 0.5
            run.add_picture(img_stream, width=Inches(display_w))
        except Exception as e:
            import traceback
            print(f"      [خطأ] فشل إضافة صورة ({elem.etype.value}): {e}")
            traceback.print_exc()
            para = doc.add_paragraph(f"[فشل استخراج صورة: {e}]")

    # ── جدول ─────────────────────────────────────────────────────────────────
    elif elem.etype == ElementType.TABLE:
        data = elem.content
        if not data:
            return
        # تنظيف البيانات
        rows = [[str(c or "") for c in row] for row in data]
        if not rows or not rows[0]:
            return

        ncols = max(len(r) for r in rows)
        nrows = len(rows)

        tbl = doc.add_table(rows=nrows, cols=ncols)
        tbl.style = "Table Grid"

        for ri, row in enumerate(rows):
            for ci, cell_txt in enumerate(row):
                if ci >= ncols:
                    break
                cell = tbl.cell(ri, ci)
                cell.text = cell_txt
                # تمييز الصف الأول (رأس الجدول)
                if ri == 0:
                    for run in cell.paragraphs[0].runs:
                        run.bold = True

        doc.add_paragraph()   # سطر فراغ بعد الجدول


# ── الدالة الرئيسية ───────────────────────────────────────────────────────────

def convert_pdf_to_docx(input_pdf: str, output_docx: str,
                         save_preview: bool = True,
                         preview_dir: str = None):
    """
    يحوّل ملف PDF إلى DOCX مع الحفاظ على التخطيط المكاني.
    """
    input_pdf  = str(input_pdf)
    output_docx = str(output_docx)

    print(f"\n{'='*60}")
    print(f"  📄 تحليل: {input_pdf}")
    print(f"{'='*60}")

    fitz_doc   = fitz.open(input_pdf)
    doc        = Document()

    # ── إعدادات الصفحة (A4 افتراضي) ────────────────────────────────────────
    section = doc.sections[0]
    section.page_width  = Cm(21)
    section.page_height = Cm(29.7)
    margin = Cm(2.0)
    section.top_margin    = margin
    section.bottom_margin = margin
    section.left_margin   = margin
    section.right_margin  = margin
    content_w_in = (21 - 4) / 2.54   # عرض المحتوى بالإنش

    preview_paths = []

    with pdfplumber.open(input_pdf) as plumber_doc:
        for page_num in range(len(fitz_doc)):
            fitz_page    = fitz_doc[page_num]
            plumber_page = plumber_doc.pages[page_num]

            pw = float(fitz_page.rect.width)
            ph = float(fitz_page.rect.height)

            print(f"\n  ── صفحة {page_num+1} ({pw:.0f}×{ph:.0f} pt) ──")

            # استخراج العناصر
            elements = extract_elements(fitz_page, plumber_page)

            counts = {}
            for e in elements:
                counts[e.etype.name] = counts.get(e.etype.name, 0) + 1
            for k, v in counts.items():
                print(f"      {k}: {v}")

            # صورة المعاينة مع المستطيلات
            if save_preview:
                ann_img = annotate_page_image(fitz_page, elements, scale=2.0)
                if preview_dir:
                    os.makedirs(preview_dir, exist_ok=True)
                    p_path = os.path.join(preview_dir, f"page_{page_num+1:03d}_preview.png")
                else:
                    p_path = os.path.splitext(output_docx)[0] + f"_page{page_num+1}_preview.png"
                ann_img.save(p_path, "PNG")
                preview_paths.append(p_path)
                print(f"      💾 معاينة: {p_path}")

            # إضافة عنوان الصفحة
            if page_num > 0:
                doc.add_page_break()
            heading = doc.add_paragraph(f"صفحة {page_num+1}")
            heading.style = "Heading 2"

            # إضافة العناصر إلى DOCX
            for elem in elements:
                add_element_to_doc(doc, elem, pw, ph, content_w_in)

    # حفظ DOCX
    doc.save(output_docx)
    print(f"\n✅ تم الحفظ: {output_docx}")
    if preview_paths:
        print(f"✅ صور المعاينة: {len(preview_paths)} ملف")

    fitz_doc.close()
    return output_docx, preview_paths


# ── نقطة الدخول ──────────────────────────────────────────────────────────────

def main():
    parser = argparse.ArgumentParser(
        description="تحويل PDF إلى DOCX مع الحفاظ على التخطيط المكاني"
    )
    parser.add_argument("input",  help="ملف PDF المدخل")
    parser.add_argument("output", nargs="?", help="ملف DOCX المخرج (اختياري)")
    parser.add_argument("--no-preview", action="store_true",
                        help="إيقاف حفظ صور المعاينة")
    parser.add_argument("--preview-dir", default=None,
                        help="مجلد لحفظ صور المعاينة")
    args = parser.parse_args()

    inp  = Path(args.input)
    out  = Path(args.output) if args.output else inp.with_suffix(".docx")

    convert_pdf_to_docx(
        str(inp), str(out),
        save_preview  = not args.no_preview,
        preview_dir   = args.preview_dir,
    )


if __name__ == "__main__":
    main()
````

def is_scanned_page(fitz_page: fitz.Page) -> bool:
    """يحدد إن كانت الصفحة مجرد صورة ممسوحة (سكانر) بدون نص حقيقي."""
    text = fitz_page.get_text("text").strip()
    images = fitz_page.get_images(full=True)
    if len(text) < 15 and len(images) >= 1:
        # تحقق أن الصورة تغطي معظم الصفحة
        page_area = fitz_page.rect.width * fitz_page.rect.height
        for img_info in images:
            xref = img_info[0]
            rects = fitz_page.get_image_rects(xref)
            for r in rects:
                img_area = (r.x1 - r.x0) * (r.y1 - r.y0)
                if img_area / page_area > 0.6:
                    return True
    return len(text) < 15 and len(images) == 0  # صفحة فارغة تماماً من النص أيضاً تحتاج تحقق


def detect_picture_regions(pil_img: Image.Image, scale: float) -> List[Tuple[float, float, float, float]]:
    """
    يكتشف مناطق الصور/الرسومات داخل صفحة سكانر (وليس النص) عبر تحليل
    كثافة الحواف والتباين المحلي. يرجّع bboxes بوحدات PDF (pt).
    """
    if not CV2_AVAILABLE:
        return []

    gray = cv2.cvtColor(np.array(pil_img), cv2.COLOR_RGB2GRAY)
    h, w = gray.shape

    # حواف قوية (كاني) لتمييز الرسومات كثيفة التفاصيل عن سطور النص الرفيعة
    edges = cv2.Canny(gray, 50, 150)

    # تكثيف بسيط لربط تفاصيل الرسم الداخلية فقط (نواة صغيرة لتجنّب الالتحام بنص الكابشن)
    kernel = cv2.getStructuringElement(cv2.MORPH_RECT, (7, 7))
    dilated = cv2.dilate(edges, kernel, iterations=2)

    contours, _ = cv2.findContours(dilated, cv2.RETR_EXTERNAL, cv2.CHAIN_APPROX_SIMPLE)

    regions = []
    page_area = w * h
    for cnt in contours:
        x, y, cw, ch = cv2.boundingRect(cnt)
        area = cw * ch
        area_ratio = area / page_area
        if area_ratio < 0.03 or area_ratio > 0.85:
            continue
        aspect = cw / max(ch, 1)
        if aspect > 4.5 or aspect < 0.2:
            continue
        roi_edges = edges[y:y+ch, x:x+cw]
        edge_density = np.count_nonzero(roi_edges) / max(area, 1)
        if edge_density < 0.06:
            continue

        # تقليم الحافة السفلية بدقة: نتحرك من الأسفل لأعلى ونوقف عند أول صف
        # كثيف الحواف بعد عبور صف فارغ كامل (هامش حقيقي بين الرسم والكابشن)
        row_density = (edges[y:y+ch, x:x+cw] > 0).sum(axis=1) / max(cw, 1)
        new_ch = ch
        i = ch - 1
        seen_content = False
        gap_count = 0
        while i >= 0:
            if row_density[i] > 0.02:
                seen_content = True
                gap_count = 0
            elif seen_content:
                gap_count += 1
                if gap_count > max(3, int(0.01 * h)):
                    new_ch = i
                    break
            i -= 1
        ch = max(new_ch, int((ch) * 0.3))  # حد أدنى أماني

        regions.append((x/scale, y/scale, (x+cw)/scale, (y+ch)/scale))

    # دمج المناطق المتداخلة/المتقاربة
    merged = []
    for r in sorted(regions, key=lambda r: -(r[2]-r[0])*(r[3]-r[1])):
        keep = True
        for m in merged:
            ix0, iy0 = max(r[0], m[0]), max(r[1], m[1])
            ix1, iy1 = min(r[2], m[2]), min(r[3], m[3])
            if ix1 > ix0 and iy1 > iy0:
                inter = (ix1-ix0)*(iy1-iy0)
                if inter / min((r[2]-r[0])*(r[3]-r[1]), (m[2]-m[0])*(m[3]-m[1])) > 0.4:
                    keep = False
                    break
        if keep:
            merged.append(r)
    return merged


def ocr_extract_elements(fitz_page: fitz.Page, lang: str = "ara+eng",
                          scale: float = 3.0) -> List["PageElement"]:
    """يستخرج عناصر نصية وصورية من صفحة ممسوحة (سكانر) باستخدام OCR + كشف بصري."""
    mat = fitz.Matrix(scale, scale)
    pix = fitz_page.get_pixmap(matrix=mat, alpha=False)
    img = Image.frombytes("RGB", [pix.width, pix.height], pix.samples)

    elements: List[PageElement] = []

    # ── 1. كشف مناطق الصور/الرسومات بصرياً ─────────────────────────────────
    picture_regions = detect_picture_regions(img, scale)
    for (px0, py0, px1, py1) in picture_regions:
        cx0, cy0 = int(px0*scale), int(py0*scale)
        cx1, cy1 = int(px1*scale), int(py1*scale)
        if cx1 - cx0 < 5 or cy1 - cy0 < 5:
            print(f"      [تحذير] منطقة صورة مكتشفة بأبعاد غير صالحة ({cx1-cx0}×{cy1-cy0}) — تم تجاهلها")
            continue
        try:
            crop = img.crop((cx0, cy0, cx1, cy1))
            buf = io.BytesIO()
            crop.save(buf, format="PNG")
            img_data = buf.getvalue()
            if len(img_data) < 100:
                print(f"      [تحذير] بيانات صورة صغيرة جداً ({len(img_data)} بايت) — تم تجاهلها")
                continue
            elements.append(PageElement(ElementType.IMAGE, (px0, py0, px1, py1), img_data))
        except Exception as e:
            print(f"      [خطأ] فشل قص/حفظ منطقة صورة: {e}")

    if not OCR_AVAILABLE:
        print("  [تحذير] pytesseract غير متاح؛ لن يتم استخراج النص من الصفحات الممسوحة")
        print("          ثبّت: pip install pytesseract --break-system-packages && apt-get install tesseract-ocr tesseract-ocr-ara")
        elements.sort(key=lambda e: (e.bbox[1], e.bbox[0]))
        return elements

    try:
        ocr_data = pytesseract.image_to_data(
            img, lang=lang, output_type=pytesseract.Output.DICT,
            config="--psm 3"
        )
    except Exception as e:
        print(f"  [خطأ OCR] {e}")
        elements.sort(key=lambda e: (e.bbox[1], e.bbox[0]))
        return elements

    def _inside_picture(x0, y0, x1, y1) -> bool:
        cx, cy = (x0+x1)/2, (y0+y1)/2
        for (px0, py0, px1, py1) in picture_regions:
            if px0 <= cx <= px1 and py0 <= cy <= py1:
                return True
        return False

    n = len(ocr_data["text"])
    # تجميع الكلمات حسب رقم الكتلة + الفقرة + السطر لإعادة بناء أسطر كاملة
    lines = {}
    for i in range(n):
        word = ocr_data["text"][i].strip()
        conf_raw = str(ocr_data.get("conf", ["-1"]*n)[i])
        conf = int(conf_raw) if conf_raw.lstrip("-").isdigit() else -1
        if not word or conf < 25:
            continue
        x, y, w, h = (ocr_data["left"][i], ocr_data["top"][i],
                      ocr_data["width"][i], ocr_data["height"][i])
        # استبعاد الكلمات الواقعة داخل منطقة صورة مكتشفة (غالباً ضجيج OCR على الرسم)
        if _inside_picture(x/scale, y/scale, (x+w)/scale, (y+h)/scale):
            continue
        key = (ocr_data["block_num"][i], ocr_data["par_num"][i], ocr_data["line_num"][i])
        if key not in lines:
            lines[key] = {"words": [], "x0": x, "y0": y, "x1": x+w, "y1": y+h}
        ld = lines[key]
        ld["words"].append(word)
        ld["x0"] = min(ld["x0"], x)
        ld["y0"] = min(ld["y0"], y)
        ld["x1"] = max(ld["x1"], x + w)
        ld["y1"] = max(ld["y1"], y + h)

    text_elements = []
    for key in sorted(lines.keys()):
        ld = lines[key]
        text = " ".join(ld["words"])
        # تحويل من بكسلات الصورة (scale×) إلى وحدات PDF (pt)
        bbox = (ld["x0"]/scale, ld["y0"]/scale, ld["x1"]/scale, ld["y1"]/scale)
        avg_h_pt = (ld["y1"] - ld["y0"]) / scale
        font_size = max(7, min(28, avg_h_pt * 0.8))
        text_elements.append(PageElement(ElementType.TEXT, bbox, text,
                                    {"avg_size": font_size, "ocr": True}))

    text_elements.sort(key=lambda e: (e.bbox[1], e.bbox[0]))

    # دمج الأسطر المتتالية القريبة عمودياً ومتشابهة بحجم الخط في فقرة واحدة
    merged: List[PageElement] = []
    for elem in text_elements:
        if merged:
            prev = merged[-1]
            same_size = abs(prev.font_info.get("avg_size", 12) - elem.font_info.get("avg_size", 12)) < 2
            vertical_gap = elem.bbox[1] - prev.bbox[3]
            line_h = prev.bbox[3] - prev.bbox[1]
            close_enough = 0 <= vertical_gap < line_h * 0.9
            similar_left = abs(elem.bbox[0] - prev.bbox[0]) < 30
            if same_size and close_enough and similar_left:
                prev.content = prev.content.rstrip() + " " + elem.content
                prev.bbox = (min(prev.bbox[0], elem.bbox[0]), prev.bbox[1],
                            max(prev.bbox[2], elem.bbox[2]), elem.bbox[3])
                continue
        merged.append(elem)

    all_elements = elements + merged
    all_elements.sort(key=lambda e: (round(e.bbox[1] / 8), e.bbox[0]))
    return all_elements


def extract_elements(fitz_page: fitz.Page, plumber_page) -> List[PageElement]:
    """يستخرج جميع العناصر من الصفحة، مع تفرّع آلي لـ OCR إن كانت الصفحة ممسوحة."""
    if is_scanned_page(fitz_page):
        print("      🔍 صفحة ممسوحة (سكانر) — تشغيل OCR + كشف الصور...")
        return ocr_extract_elements(fitz_page)
    return _extract_elements_native(fitz_page, plumber_page)


def _extract_elements_native(fitz_page: fitz.Page, plumber_page) -> List[PageElement]:
    """يستخرج جميع العناصر من الصفحة ويصنّفها."""
    elements: List[PageElement] = []
    used_rects = []   # مناطق مستخدمة لتجنب التداخل

    # ── 1. الجداول (pdfplumber أفضل للجداول) ──────────────────────────────
    try:
        tables = plumber_page.extract_tables(
            {"vertical_strategy": "lines", "horizontal_strategy": "lines"}
        )
        table_bboxes = []
        for t_obj in plumber_page.find_tables():
            bb = t_obj.bbox   # (x0,top,x1,bottom)  pdfplumber coords
            # تحويل إحداثيات pdfplumber → fitz  (y محور مقلوب)
            page_h = float(plumber_page.height)
            fx0, fy0 = bb[0], page_h - bb[3]
            fx1, fy1 = bb[2], page_h - bb[1]
            table_bboxes.append((fx0, fy0, fx1, fy1))
            data = t_obj.extract()
            elements.append(PageElement(ElementType.TABLE, (fx0,fy0,fx1,fy1), data))
            used_rects.append(fitz.Rect(fx0, fy0, fx1, fy1))
    except Exception as e:
        print(f"  [تحذير] استخراج الجداول: {e}")

    # ── 2. الصور المضمّنة ────────────────────────────────────────────────────
    img_list = fitz_page.get_images(full=True)
    for img_info in img_list:
        xref = img_info[0]
        # الحصول على bbox الصورة داخل الصفحة
        img_rects = fitz_page.get_image_rects(xref)
        if not img_rects:
            continue
        r = img_rects[0]
        # تجاهل صور الخلفية الكبيرة جداً (>85% من الصفحة)
        page_area = fitz_page.rect.width * fitz_page.rect.height
        img_area  = (r.x1 - r.x0) * (r.y1 - r.y0)
        if img_area / page_area > 0.85:
            continue
        # استخراج بيانات الصورة
        try:
            base_img = fitz_page.parent.extract_image(xref)
            img_bytes = base_img["image"]
            bbox = (r.x0, r.y0, r.x1, r.y1)

            # هل هي رسم بياني؟ نفحص بالحجم والموقع
            etype = ElementType.IMAGE
            # رسوم بيانية عادةً: عرض > 200pt وارتفاع > 150pt
            if (r.x1-r.x0) > 200 and (r.y1-r.y0) > 150:
                etype = ElementType.CHART

            elements.append(PageElement(etype, bbox, img_bytes))
            used_rects.append(r)
        except Exception:
            pass

    # ── 3. كتل النصوص ───────────────────────────────────────────────────────
    blocks = fitz_page.get_text("dict", flags=fitz.TEXT_PRESERVE_WHITESPACE)["blocks"]
    for blk in blocks:
        if blk["type"] != 0:   # 0 = نص
            continue
        r = fitz.Rect(blk["bbox"])
        # تحقق من أن الكتلة لا تتداخل مع جدول/صورة موجودة
        overlap = any(r.intersects(ur) and _overlap_ratio(r, ur) > 0.5 for ur in used_rects)
        if overlap:
            continue

        # تجميع معلومات الخط
        lines_text = []
        font_sizes = []
        for line in blk.get("lines", []):
            for span in line.get("spans", []):
                lines_text.append(span["text"])
                font_sizes.append(span["size"])

        full_text = "\n".join(lines_text).strip()
        if not full_text:
            continue

        avg_size = sum(font_sizes) / len(font_sizes) if font_sizes else 12
        bbox = (r.x0, r.y0, r.x1, r.y1)
        elem = PageElement(ElementType.TEXT, bbox, full_text,
                           {"avg_size": avg_size, "spans": blk})
        elements.append(elem)

    # ── ترتيب العناصر من أعلى لأسفل ثم يسار لأيمن ─────────────────────────
    elements.sort(key=lambda e: (round(e.bbox[1] / 10), e.bbox[0]))
    return elements


def _overlap_ratio(r1: fitz.Rect, r2: fitz.Rect) -> float:
    inter = r1 & r2
    if inter.is_empty:
        return 0.0
    return inter.get_area() / min(r1.get_area(), r2.get_area() + 1e-9)

def annotate_page_image(fitz_page: fitz.Page, elements: List[PageElement],
                        scale: float = 2.0) -> Image.Image:
    """يرسم صورة للصفحة مع مستطيلات ترميزية حول كل عنصر."""
    mat = fitz.Matrix(scale, scale)
    pix = fitz_page.get_pixmap(matrix=mat, alpha=False)
    img = Image.frombytes("RGB", [pix.width, pix.height], pix.samples)
    draw = ImageDraw.Draw(img, "RGBA")

    try:
        font = ImageFont.truetype("/usr/share/fonts/truetype/dejavu/DejaVuSans-Bold.ttf", 14)
    except Exception:
        font = ImageFont.load_default()

    for elem in elements:
        x0, y0, x1, y1 = [c * scale for c in elem.bbox]
        r, g, b = ELEMENT_COLORS[elem.etype]
        # مستطيل شبه شفاف
        draw.rectangle([x0, y0, x1, y1],
                       outline=(r, g, b, 255),
                       fill=(r, g, b, 25),
                       width=2)
        # تسمية في أعلى اليسار
        label = ELEMENT_LABELS[elem.etype]
        tx, ty = x0 + 3, y0 + 2
        # خلفية للنص
        draw.rectangle([tx - 1, ty - 1, tx + len(label)*7 + 2, ty + 16],
                       fill=(r, g, b, 180))
        draw.text((tx, ty), label, fill="white", font=font)

    return img

def _set_cell_border(cell, **kwargs):
    """ضبط حدود خلية جدول."""
    tc = cell._tc
    tcPr = tc.get_or_add_tcPr()
    tcBorders = OxmlElement("w:tcBorders")
    for edge in ("top", "left", "bottom", "right", "insideH", "insideV"):
        edge_data = kwargs.get(edge)
        if edge_data:
            tag = OxmlElement(f"w:{edge}")
            for attr, val in edge_data.items():
                tag.set(qn(f"w:{attr}"), str(val))
            tcBorders.append(tag)
    tcPr.append(tcBorders)


def _contains_arabic(text: str) -> bool:
    return any('\u0600' <= ch <= '\u06FF' for ch in text)


def _set_rtl(paragraph):
    """يضبط اتجاه الفقرة من اليمين لليسار (للنصوص العربية)."""
    pPr = paragraph._p.get_or_add_pPr()
    bidi = OxmlElement("w:bidi")
    pPr.append(bidi)


def add_element_to_doc(doc: Document, elem: PageElement,
                        page_w_pt: float, page_h_pt: float,
                        content_w_in: float):
    """يضيف عنصراً واحداً إلى مستند الـ DOCX مع الحفاظ على نسبه المكانية."""

    x0, y0, x1, y1 = elem.bbox
    # نسبة العرض
    rel_x   = x0 / page_w_pt          # من أين يبدأ (0→1)
    rel_w   = (x1 - x0) / page_w_pt   # عرضه النسبي
    elem_w_in = content_w_in * rel_w

    # ── نص ──────────────────────────────────────────────────────────────────
    if elem.etype == ElementType.TEXT:
        para = doc.add_paragraph()
        run  = para.add_run(elem.content)
        avg_sz = elem.font_info.get("avg_size", 12)
        run.font.size = Pt(max(6, avg_sz))

        is_arabic = _contains_arabic(elem.content)
        if is_arabic:
            run.font.name = "Arial"
            _set_rtl(para)
            # محاذاة معكوسة للعربية: يمين = بداية النص
            if rel_x > 0.55:
                para.alignment = WD_ALIGN_PARAGRAPH.LEFT
            elif rel_x > 0.25:
                para.alignment = WD_ALIGN_PARAGRAPH.CENTER
            else:
                para.alignment = WD_ALIGN_PARAGRAPH.RIGHT
        else:
            if rel_x > 0.55:
                para.alignment = WD_ALIGN_PARAGRAPH.RIGHT
            elif rel_x > 0.25:
                para.alignment = WD_ALIGN_PARAGRAPH.CENTER
            else:
                para.alignment = WD_ALIGN_PARAGRAPH.LEFT
        # مسافة بادئة بسيطة
        left_indent = Inches(content_w_in * rel_x * 0.7)
        para.paragraph_format.left_indent = min(left_indent, Inches(content_w_in * 0.5))

    # ── صورة / رسم بياني ─────────────────────────────────────────────────────
    elif elem.etype in (ElementType.IMAGE, ElementType.CHART):
        try:
            img_bytes = elem.content
            if not img_bytes:
                print(f"      [تحذير] عنصر صورة بدون بيانات — تم تجاهله")
                return
            img_stream = io.BytesIO(img_bytes)
            # تحقق من صحة الصورة بدون استخدام verify() الذي يُفسد الـ stream في بعض إصدارات Pillow
            pil_img = Image.open(img_stream)
            pil_img.load()          # يقرأ البيانات فعلياً للتأكد من سلامتها
            img_stream.seek(0)      # نعيد المؤشر لأول الستريم بعد القراءة

            para = doc.add_paragraph()
            if rel_x > 0.55:
                para.alignment = WD_ALIGN_PARAGRAPH.RIGHT
            elif rel_x > 0.25:
                para.alignment = WD_ALIGN_PARAGRAPH.CENTER
            else:
                para.alignment = WD_ALIGN_PARAGRAPH.LEFT

            run = para.add_run()
            display_w = min(elem_w_in, content_w_in)
            if display_w <= 0:
                display_w = content_w_in * 0.5
            run.add_picture(img_stream, width=Inches(display_w))
        except Exception as e:
            import traceback
            print(f"      [خطأ] فشل إضافة صورة ({elem.etype.value}): {e}")
            traceback.print_exc()
            para = doc.add_paragraph(f"[فشل استخراج صورة: {e}]")

    # ── جدول ─────────────────────────────────────────────────────────────────
    elif elem.etype == ElementType.TABLE:
        data = elem.content
        if not data:
            return
        # تنظيف البيانات
        rows = [[str(c or "") for c in row] for row in data]
        if not rows or not rows[0]:
            return

        ncols = max(len(r) for r in rows)
        nrows = len(rows)

        tbl = doc.add_table(rows=nrows, cols=ncols)
        tbl.style = "Table Grid"

        for ri, row in enumerate(rows):
            for ci, cell_txt in enumerate(row):
                if ci >= ncols:
                    break
                cell = tbl.cell(ri, ci)
                cell.text = cell_txt
                # تمييز الصف الأول (رأس الجدول)
                if ri == 0:
                    for run in cell.paragraphs[0].runs:
                        run.bold = True

        doc.add_paragraph()   # سطر فراغ بعد الجدول

def convert_pdf_to_docx(input_pdf: str, output_docx: str,
                         save_preview: bool = True,
                         preview_dir: str = None):
    """
    يحوّل ملف PDF إلى DOCX مع الحفاظ على التخطيط المكاني.
    """
    input_pdf  = str(input_pdf)
    output_docx = str(output_docx)

    print(f"\n{'='*60}")
    print(f"  📄 تحليل: {input_pdf}")
    print(f"{'='*60}")

    fitz_doc   = fitz.open(input_pdf)
    doc        = Document()

    # ── إعدادات الصفحة (A4 افتراضي) ────────────────────────────────────────
    section = doc.sections[0]
    section.page_width  = Cm(21)
    section.page_height = Cm(29.7)
    margin = Cm(2.0)
    section.top_margin    = margin
    section.bottom_margin = margin
    section.left_margin   = margin
    section.right_margin  = margin
    content_w_in = (21 - 4) / 2.54   # عرض المحتوى بالإنش

    preview_paths = []

    with pdfplumber.open(input_pdf) as plumber_doc:
        for page_num in range(len(fitz_doc)):
            fitz_page    = fitz_doc[page_num]
            plumber_page = plumber_doc.pages[page_num]

            pw = float(fitz_page.rect.width)
            ph = float(fitz_page.rect.height)

            print(f"\n  ── صفحة {page_num+1} ({pw:.0f}×{ph:.0f} pt) ──")

            # استخراج العناصر
            elements = extract_elements(fitz_page, plumber_page)

            counts = {}
            for e in elements:
                counts[e.etype.name] = counts.get(e.etype.name, 0) + 1
            for k, v in counts.items():
                print(f"      {k}: {v}")

            # صورة المعاينة مع المستطيلات
            if save_preview:
                ann_img = annotate_page_image(fitz_page, elements, scale=2.0)
                if preview_dir:
                    os.makedirs(preview_dir, exist_ok=True)
                    p_path = os.path.join(preview_dir, f"page_{page_num+1:03d}_preview.png")
                else:
                    p_path = os.path.splitext(output_docx)[0] + f"_page{page_num+1}_preview.png"
                ann_img.save(p_path, "PNG")
                preview_paths.append(p_path)
                print(f"      💾 معاينة: {p_path}")

            # إضافة عنوان الصفحة
            if page_num > 0:
                doc.add_page_break()
            heading = doc.add_paragraph(f"صفحة {page_num+1}")
            heading.style = "Heading 2"

            # إضافة العناصر إلى DOCX
            for elem in elements:
                add_element_to_doc(doc, elem, pw, ph, content_w_in)

    # حفظ DOCX
    doc.save(output_docx)
    print(f"\n✅ تم الحفظ: {output_docx}")
    if preview_paths:
        print(f"✅ صور المعاينة: {len(preview_paths)} ملف")

    fitz_doc.close()
    return output_docx, preview_paths


# ── نقطة الدخول ──────────────────────────────────────────────────────────────

def main():
    parser = argparse.ArgumentParser(
        description="تحويل PDF إلى DOCX مع الحفاظ على التخطيط المكاني"
    )
    parser.add_argument("input",  help="ملف PDF المدخل")
    parser.add_argument("output", nargs="?", help="ملف DOCX المخرج (اختياري)")
    parser.add_argument("--no-preview", action="store_true",
                        help="إيقاف حفظ صور المعاينة")
    parser.add_argument("--preview-dir", default=None,
                        help="مجلد لحفظ صور المعاينة")
    args = parser.parse_args()

    inp  = Path(args.input)
    out  = Path(args.output) if args.output else inp.with_suffix(".docx")

    convert_pdf_to_docx(
        str(inp), str(out),
        save_preview  = not args.no_preview,
        preview_dir   = args.preview_dir,
    )


if __name__ == "__main__":
    main()

نعم، موجود أدوات مجانية ومفتوحة المصدر بتعمل layout analysis حقيقي، بمستويات جودة مختلفة. هذا ملخص للأهم منها:

### اللي السكريبت بيعمله بالترتيب:
1. يشغّل OCR ويرسم المستطيلات الملونة (نفس الكود بتاعك بدون تغيير)
2. يجمّع الكلمات إلى أسطر، والأسطر التابعة لنفس الفقرة (`block_num` + `par_num`) إلى فقرة واحدة متصلة
3. يكتب كل فقرة كـ `paragraph` منفصل في ملف Word
4. يحفظ الملف في `/content/extracted_text.docx`
5. يطبع معاينة سريعة للنص في الكولاب نفسه

ثم شغّل السكريبت المرفق مباشرة.

**ملاحظة:** السكريبت ده بيستخرج النص فقط (مش الصورة)، لأن طلبك كان "نسخ المستخرج كله". لو عايز كمان الصورة (الرسم) تتضاف لملف الـ docx بجانب النص، قولّي وأضيفها — هي نفس آلية كشف الصورة اللي عندنا في البرنامج الأساسي.

````python
"""
يشغّل OCR على صورة الصفحة، يرسم مستطيلات حول كل كتلة نص (للمعاينة)،
ثم يصدّر كل النص المستخرج (مُجمّعاً حسب الفقرة/السطر) إلى ملف Word (.docx)
"""

import pytesseract
from PIL import Image, ImageDraw
from docx import Document
from docx.shared import Pt

# ── 1. تحميل الصورة ──────────────────────────────────────────────────────
img_path = "/content/Screenshot from 2026-06-30 02-20-37.png"
img = Image.open(img_path).convert("RGB")
draw = ImageDraw.Draw(img)

# ── 2. تشغيل OCR واستخراج بيانات كل كلمة (موقع + نص + رقم الكتلة/السطر) ──
data = pytesseract.image_to_data(
    img, lang="ara+eng", config="--psm 3",
    output_type=pytesseract.Output.DICT
)

n_boxes = len(data['text'])

# ── 3. رسم مستطيل حول كل كلمة، ولوّن بحسب رقم الـ block ──────────────────
for i in range(n_boxes):
    if int(data['conf'][i]) > 30 and data['text'][i].strip():
        x, y, w, h = data['left'][i], data['top'][i], data['width'][i], data['height'][i]
        block_num = data['block_num'][i]
        color = ["red", "blue", "green", "orange", "purple"][block_num % 5]
        draw.rectangle([x, y, x + w, y + h], outline=color, width=2)

img.save("layout_analysis_result.png")
from IPython.display import display
display(img)

# ── 4. تجميع الكلمات إلى أسطر، ثم الأسطر إلى فقرات (حسب block_num/par_num) ──
lines = {}   # key = (block_num, par_num, line_num)
for i in range(n_boxes):
    word = data['text'][i].strip()
    conf = int(data['conf'][i]) if str(data['conf'][i]).lstrip("-").isdigit() else -1
    if not word or conf <= 30:
        continue
    key = (data['block_num'][i], data['par_num'][i], data['line_num'][i])
    lines.setdefault(key, []).append(word)

# دمج الأسطر التابعة لنفس الفقرة (نفس block_num + par_num) في فقرة واحدة
paragraphs = {}   # key = (block_num, par_num)
for (block_num, par_num, line_num), words in sorted(lines.items()):
    p_key = (block_num, par_num)
    line_text = " ".join(words)
    paragraphs.setdefault(p_key, []).append(line_text)

# ── 5. كتابة كل فقرة إلى ملف Word ────────────────────────────────────────
doc = Document()
doc.styles["Normal"].font.name = "Arial"
doc.styles["Normal"].font.size = Pt(12)

for p_key in sorted(paragraphs.keys()):
    lines_in_para = paragraphs[p_key]
    full_text = " ".join(lines_in_para).strip()
    if not full_text:
        continue
    doc.add_paragraph(full_text)

output_path = "/content/extracted_text.docx"
doc.save(output_path)
print(f"✅ تم حفظ النص المستخرج في: {output_path}")

# عرض سريع للنص النهائي في الكولاب للمراجعة
print("\n--- معاينة النص المستخرج ---")
for p_key in sorted(paragraphs.keys()):
    print(" ".join(paragraphs[p_key]))
````

Let's check the detected picture region bounds precisely:المنطقة المكتشفة بتمتد لـ y=567، والكابشن النصي على الأغلب يبدأ قريب من هذا الحد (لأن الصورة الصغيرة 919px مضغوطة، فالفجوة بين الصورة والكابشن أصغر بكثير من النسخة الأصلية بدقة 300dpi). الـ trimming algorithm يحتاج معايرة بحسب الدقة (scale-aware)، فالقيم الثابتة (`gap_count > max(3, int(0.01*h))`) بتشتغل صح على صور كبيرة لكن صارمة جداً على صور صغيرة. دعني أصلح هذا بجعل المعايير نسبية لدقة الصورة:لازال 4 عناصر فقط. دعني أتحقق من المنطقة المكتشفة الجديدة وأرى الصورة المرسومة:المنطقة لسة وقفت عند y=573، يعني خوارزمية الـ trimming لسة شايفة الكابشن جزء من كثافة الحواف. السبب الحقيقي: الكابشن النصي نفسه عنده كثافة حواف عالية (لأنه bold/منمّق) فهو بيبان "كمحتوى" مش "فجوة"، فالخوارزمية ما بتوقفش قبله. خليني أفحص بصرياً:الآن أرى المشكلة بوضوح: في y=542-544 يوجد خط حدود الصورة (كثافة عالية 0.57)، ثم فجوة حقيقية (y=546-550)، **لكن بعدها مباشرة كثافة عالية تاني (y=556-564 ≈ 0.3-0.5) — وهذا هو الكابشن نفسه**، لأن الكابشن مكتوب bold فكثافته عالية جداً. خوارزمية "أول فجوة" بتوقف صحيح عند الفجوة الأولى الحقيقية الموجودة في y≈546-552، لكن النتيجة كانت 573 بمعنى إنها ما توقفتش هناك. يبدو المشكلة أن `gap_count > min_gap` (حيث `min_gap=2` للصورة الصغيرة دي) لسة كبير جداً نسبة لحجم الفجوة الفعلية (حوالي 5 صفوف فقط من y=546-551).

دعني أجرب قيمة أصغر:`min_gap=3`، والفجوة الفعلية بين y=546 و y=552 فيها فقط حوالي صفين بكثافة منخفضة جداً قبل ما الكابشن يبدأ (لاحظ الانتقال السريع جداً 0→0.029→0.1→0.375). يعني المسافة الحقيقية صغيرة جداً (الصورة مضغوطة بدقة منخفضة)، فمعيار "3 صفوف فجوة" أكبر من الفجوة الفعلية المتاحة. هذا النوع من التمييز (صورة قريبة جداً من نص bold) صعب جداً بالاعتماد فقط على كثافة الحواف وحدها بدون معلومة إضافية. الحل الأشمل هو الاعتماد على **حد الصورة الفعلي (الإطار/border)** بدل التخمين بالفجوة:الآن استعدنا الكابشن! 5 عناصر: صورة + 4 نصوص (بما فيها الكابشن المفقود). دعني أتحقق بصرياً من النتيجة النهائية والترتيب الكامل:الترتيب الآن صحيح تماماً ومطابق للصفحة الأصلية: الصورة أولاً، ثم الكابشن، ثم العنوان، إلخ. دعني أتحقق بصرياً من المعاينة بالمستطيلات للتأكد من الدقة النهائية:

````python
"""
يشغّل OCR على صورة الصفحة، يستخرج النصوص والرسومات والجداول
يرسم مستطيلات حول كل كتلة (للمعاينة)،
ثم يصدّر كل شيء إلى ملف Word (.docx) مع الحفاظ على التخطيط
"""

import pytesseract
import cv2
import numpy as np
from PIL import Image, ImageDraw
from docx import Document
from docx.shared import Pt, Inches
from docx.enum.text import WD_ALIGN_PARAGRAPH
import io

# ── 1. تحميل الصورة ──────────────────────────────────────────────────────
img_path = "/content/Screenshot from 2026-06-30 02-20-37.png"
pil_img = Image.open(img_path).convert("RGB")

# تحويل الصورة إلى صيغة OpenCV للتعامل معها ك مصفوفة بكسلات
img_cv = cv2.cvtColor(np.array(pil_img), cv2.COLOR_RGB2BGR)
h, w = img_cv.shape[:2]

# نسخة للمعاينة (رسم المستطيلات)
draw_img = pil_img.copy()
draw = ImageDraw.Draw(draw_img)

# ── 2. تشغيل OCR واستخراج بيانات كل كلمة ────────────────────────────────
# استخدام eng فقط لأن الكتاب إنجليزي (أسرع وأدق)
data = pytesseract.image_to_data(
    pil_img, lang="eng", config="--psm 3",
    output_type=pytesseract.Output.DICT
)

n_boxes = len(data['text'])

# ── 3. تجميع الكلمات إلى فقرات مع حساب الإحداثيات لكل فقرة ─────────────
lines = {}
for i in range(n_boxes):
    word = data['text'][i].strip()
    conf = int(data['conf'][i]) if str(data['conf'][i]).lstrip("-").isdigit() else -1
    if not word or conf <= 30:
        continue

    x, y, bw, bh = data['left'][i], data['top'][i], data['width'][i], data['height'][i]
    key = (data['block_num'][i], data['par_num'][i], data['line_num'][i])

    # حفظ الكلمة مع موقعها
    lines.setdefault(key, []).append({'word': word, 'x': x, 'y': y, 'w': bw, 'h': bh})

paragraphs = {}
for (block_num, par_num, line_num), words_info in sorted(lines.items()):
    p_key = (block_num, par_num)

    # حساب المربع المحيط بالفقرة بأكملها
    min_x = min(w_info['x'] for w_info in words_info)
    min_y = min(w_info['y'] for w_info in words_info)
    max_x = max(w_info['x'] + w_info['w'] for w_info in words_info)
    max_y = max(w_info['y'] + w_info['h'] for w_info in words_info)

    full_text = " ".join([w_info['word'] for w_info in words_info])

    paragraphs[p_key] = {
        'text': full_text,
        'bbox': [min_x, min_y, max_x, max_y],
        'center_y': (min_y + max_y) / 2
    }

    # رسم المستطيل الأزرق حول النصوص في صورة المعاينة
    draw.rectangle([min_x, min_y, max_x, max_y], outline="blue", width=2)

# ── 4. اكتشاف الرسومات والجداول (المناطق الخالية من النص) ──────────────
mask = np.ones((h, w), dtype=np.uint8) * 255
for p_info in paragraphs.values():
    x, y, x2, y2 = p_info['bbox']
    # رسم مناطق النصوص بالأسود على القناع الأبيض
    cv2.rectangle(mask, (max(0, x-5), max(0, y-5)), (min(w, x2+5), min(h, y2+5)), 0, -1)

# البحث عن المساحات البيضاء المتبقية (هي الرسومات والصور)
contours, _ = cv2.findContours(mask, cv2.RETR_EXTERNAL, cv2.CHAIN_APPROX_SIMPLE)
image_regions = []

for cnt in contours:
    cx, cy, cw, ch = cv2.boundingRect(cnt)
    area = cw * ch

    # تجاهل الحبيبات الصغيرة جداً
    if cw < 40 or ch < 40:
        continue
    # تجاهل خلفية الصفحة الكبيرة
    if area > (h * w * 0.85):
        continue

    image_regions.append({
        'bbox': [cx, cy, cx + cw, cy + ch],
        'center_y': (cy + cy + ch) / 2
    })

    # رسم المستطيل الأخضر حول الرسومات في صورة المعاينة
    draw.rectangle([cx, cy, cx + cw, cy + ch], outline="green", width=3)

# عرض المعاينة (أزرق = نص، أخضر = رسم/صورة سيتم استخراجها)
from IPython.display import display
print("المعاينة: (الأزرق للنصوص، الأخضر للرسومات/الجداول التي سيتم حفظها كصور)")
display(draw_img)

# ── 5. دمج النصوص والصور في قائمة واحدة وترتيبها ─────────────────────
all_elements = []

for p_info in paragraphs.values():
    all_elements.append({'type': 'text', **p_info})

for img_reg in image_regions:
    all_elements.append({'type': 'image', **img_reg})

# ترتيب العناصر من أعلى الصفحة لأسفلها للحفاظ على التخطيط
all_elements.sort(key=lambda x: x['center_y'])

# ── 6. كتابة كل شيء إلى ملف Word ────────────────────────────────────────
doc = Document()
doc.styles["Normal"].font.name = "Times New Roman"
doc.styles["Normal"].font.size = Pt(12)

for elem in all_elements:
    if elem['type'] == 'image':
        x1, y1, x2, y2 = elem['bbox']

        # قص منطقة الرسمة من الصورة الأصلية
        cropped_img = pil_img.crop((x1, y1, x2, y2))

        # تحويل الصورة المقصوصة إلى bytes لإدخالها للوورد
        img_byte_arr = io.BytesIO()
        cropped_img.save(img_byte_arr, format='PNG')
        img_bytes = img_byte_arr.getvalue()

        # إضافة الصورة للوورد
        para = doc.add_paragraph()
        para.alignment = WD_ALIGN_PARAGRAPH.CENTER
        run = para.add_run()

        # حساب العرض بالبوصة (افتراضياًScreenshot بدقة 150 DPI)
        img_width_inches = (x2 - x1) / 150.0
        if img_width_inches > 0.5:
            run.add_picture(io.BytesIO(img_bytes), width=Inches(img_width_inches))

    elif elem['type'] == 'text':
        # التأكد من أن النص ليس داخل صورة (تجنب تكرار النص إذا كان على الرسمة)
        is_inside_image = False
        for img_reg in image_regions:
            ix1, iy1, ix2, iy2 = img_reg['bbox']
            x_overlap = max(0, min(elem['bbox'][2], ix2) - max(elem['bbox'][0], ix1))
            y_overlap = max(0, min(elem['bbox'][3], iy2) - max(elem['bbox'][1], iy1))
            overlap_area = x_overlap * y_overlap
            elem_area = (elem['bbox'][2] - elem['bbox'][0]) * (elem['bbox'][3] - elem['bbox'][1])

            # إذا تداخل النص مع الصورة بأكثر من 50%، نتجاهله (لأن الصورة تأخذه كاملاً)
            if elem_area > 0 and (overlap_area / elem_area) > 0.5:
                is_inside_image = True
                break

        if not is_inside_image:
            para = doc.add_paragraph(elem['text'])
            para.alignment = WD_ALIGN_PARAGRAPH.LEFT

output_path = "/content/extracted_all.docx"
doc.save(output_path)
print(f"\n✅ تم حفظ النصوص والرسومات معاً بنجاح في: {output_path}")
````

````python
"""
استخراج شامل: نصوص + رسومات + جداول
باستخدام الذكاء الاصطناعي لاكتشاف مواقع كل عنصر بدقة بكسل
"""

import layoutparser as lp
import pytesseract
import cv2
import numpy as np
from PIL import Image
from docx import Document
from docx.shared import Pt, Inches
from docx.enum.text import WD_ALIGN_PARAGRAPH
import io

# ── 1. تحميل نموذج الذكاء الاصطناعي لاكتشاف عناصر الصفحة ─────────────
print("جاري تحميل نموذج اكتشاف التخطيط (يحدث مرة واحدة فقط)...")
# نموذج مدرب على ملايين الصفحات لاكتشاف (نص، عنوان، صورة، جدول، قائمة)
model = lp.Detectron2LayoutModel(
    config_path='lp://PubLayNet/mask_rcnn_R_50_FPN_3x/config',
    extra_config=["MODEL.ROI_HEADS.SCORE_THRESH_TEST", 0.5], # دقة الكشف
    label_map={0: "Text", 1: "Title", 2: "List", 3: "Table", 4: "Figure"}
)
print("✅ تم التحميل بنجاح!\n")

# ── 2. تحليل الصورة ──────────────────────────────────────────────────────
img_path = "/content/Screenshot from 2026-06-30 02-20-37.png"
pil_img = Image.open(img_path).convert("RGB")
img_cv = cv2.cvtColor(np.array(pil_img), cv2.COLOR_RGB2BGR)

# الكشف عن العناصر
page_layout = model.detect(pil_img)

# ── 3. فرز العناصر من أعلى الصفحة لأسفلها ─────────────────────────────
# تحويل النتائج إلى قائمة بسيطة لنتمكن من فرزها
elements_list = []
for block in page_layout:
    x1, y1 = block.coordinates[0]
    x2, y2 = block.coordinates[2]

    elements_list.append({
        'type': block.type,       # Text, Figure, Table
        'bbox': [int(x1), int(y1), int(x2), int(y2)],
        'center_y': (y1 + y2) / 2
    })

# الترتيب العمودي
elements_list.sort(key=lambda x: x['center_y'])

# ── 4. إنشاء ملف Word وملء البيانات ────────────────────────────────────
doc = Document()
doc.styles["Normal"].font.name = "Times New Roman"
doc.styles["Normal"].font.size = Pt(12)

for elem in elements_list:
    x1, y1, x2, y2 = elem['bbox']

    # إذا كان العنصر نصاً أو قائمة أو عنوان
    if elem['type'] in ['Text', 'Title', 'List']:
        # قص منطقة النص فقط
        cropped_text_img = pil_img.crop((x1, y1, x2, y2))

        # تشغيل Tesseract على هذه المنطقة المحددة فقط (أسرع وأدق)
        text = pytesseract.image_to_string(cropped_text_img, lang="eng", config="--psm 6").strip()

        if text:
            para = doc.add_paragraph(text)
            if elem['type'] == 'Title':
                para.runs[0].bold = True
                para.runs[0].font.size = Pt(16)
            else:
                para.alignment = WD_ALIGN_PARAGRAPH.LEFT

    # إذا كان العنصر صورة أو جدول (نعامله كصورة للحفاظ على شكله)
    elif elem['type'] in ['Figure', 'Table']:
        # قص المنطقة من الصورة الأصلية
        cropped_visual_img = pil_img.crop((x1, y1, x2, y2))

        # تحويلها إلى صيغة bytes للوورد
        img_byte_arr = io.BytesIO()
        cropped_visual_img.save(img_byte_arr, format='PNG')
        img_bytes = img_byte_arr.getvalue()

        # إضافتها للوورد
        para = doc.add_paragraph()
        para.alignment = WD_ALIGN_PARAGRAPH.CENTER
        run = para.add_run()

        # حساب العرض (بافتراض دقة 150 DPI للصورة الملتقطة)
        width_inches = (x2 - x1) / 150.0
        if width_inches > 0.5:
            run.add_picture(io.BytesIO(img_bytes), width=Inches(width_inches))

# ── 5. الحفظ ─────────────────────────────────────────────────────────────
output_path = "/content/final_extracted_all.docx"
doc.save(output_path)
print(f"✅ تم حفظ كل العناصر (نصوص + رسومات + جداول) بنجاح في:\n{output_path}")

# معاينة سريعة لما تم اكتشافه
print("\n--- العناصر التي تم اكتشافها في الصفحة ---")
for i, elem in enumerate(elements_list):
    print(f"{i+1}. [{elem['type']}] الموقع: {elem['bbox']}")
````

````python


"""
الحل المضمون: استخراج النصوص والرسومات والجداول من صورة واحدة
باستخدام EasyOCR + خوارزمية اكتشاف الفراغات (المجربة والمضمونة)
"""

import cv2
import numpy as np
from PIL import Image
from docx import Document
from docx.shared import Inches, Pt
from docx.enum.text import WD_ALIGN_PARAGRAPH
import io
import logging

logging.getLogger("easyocr").setLevel(logging.ERROR)

# ── 1. تحميل الصورة ──────────────────────────────────────────────────────
img_path = "/content/Screenshot from 2026-06-30 02-20-37.png"
print("جاري تحميل الصورة...")
pil_img = Image.open(img_path).convert("RGB")
img_cv = cv2.cvtColor(np.array(pil_img), cv2.COLOR_RGB2BGR)
h, w = img_cv.shape[:2]

# ── 2. تحميل محرك OCR للإنجليزية ────────────────────────────────────────
print("جاري تحميل محرك القراءة...")
import easyocr
reader = easyocr.Reader(['en'], gpu=False)
print("✅ جاهز!\n")

# ── 3. استخراج النصوص وإحداثياتها ───────────────────────────────────────
print("⚙️ قراءة النصوص...")
results = reader.readtext(img_path)

text_boxes = []
texts = []

for (bbox, text, conf) in results:
    x_min = int(min([p[0] for p in bbox]))
    y_min = int(min([p[1] for p in bbox]))
    x_max = int(max([p[0] for p in bbox]))
    y_max = int(max([p[1] for p in bbox]))
    text_boxes.append([x_min, y_min, x_max, y_max])
    texts.append(text)

# ── 4. اكتشاف الصور والرسومات والجداول (المناطق الخالية من النص) ─────
print("⚙️ اكتشاف الرسومات والجداول...")
mask = np.ones((h, w), dtype=np.uint8) * 255
for box in text_boxes:
    x_min, y_min, x_max, y_max = box
    cv2.rectangle(mask, (x_min - 3, y_min - 3), (x_max + 3, y_max + 3), 0, -1)

contours, _ = cv2.findContours(mask, cv2.RETR_EXTERNAL, cv2.CHAIN_APPROX_SIMPLE)
image_regions = []

for cnt in contours:
    x, y, cw, ch = cv2.boundingRect(cnt)
    area = cw * ch

    # تجاهل الخلفية البيضاء الكبيرة جداً للصفحة
    if area > (h * w * 0.9):
        continue
    # تجاهل الحبيبات الصغيرة (النقاط والأخطاء)
    if cw < 50 or ch < 50:
        continue

    image_regions.append([x, y, x + cw, y + ch])

# ── كشف الجداول تحديداً (عبر الخطوط) ──────────────────────────────────
gray = cv2.cvtColor(img_cv, cv2.COLOR_BGR2GRAY)
thresh = cv2.threshold(gray, 150, 255, cv2.THRESH_BINARY_INV)[1]

hard_horizontal_kernel = cv2.getStructuringElement(cv2.MORPH_RECT, (40, 1))
hard_horizontal_lines = cv2.morphologyEx(thresh, cv2.MORPH_OPEN, hard_horizontal_kernel, iterations=2)

hard_vertical_kernel = cv2.getStructuringElement(cv2.MORPH_RECT, (1, 40))
hard_vertical_lines = cv2.morphologyEx(thresh, cv2.MORPH_OPEN, hard_vertical_kernel, iterations=2)

table_mask = cv2.add(hard_horizontal_lines, hard_vertical_lines)
table_contours, _ = cv2.findContours(table_mask, cv2.RETR_EXTERNAL, cv2.CHAIN_APPROX_SIMPLE)

for cnt in table_contours:
    x, y, cw, ch = cv2.boundingRect(cnt)
    if cw > 80 and ch > 40:
        image_regions.append([x, y, x + cw, y + ch])

# ── 5. دمج وترتيب العناصر ──────────────────────────────────────────────
elements = []
for i, box in enumerate(text_boxes):
    elements.append({'type': 'text', 'bbox': box, 'content': texts[i], 'center_y': (box[1] + box[3]) / 2})

for region in image_regions:
    elements.append({'type': 'image', 'bbox': region, 'center_y': (region[1] + region[3]) / 2})

elements.sort(key=lambda x: x['center_y'])

# ── 6. إنشاء ملف Word ──────────────────────────────────────────────────
print("⚙️ إنشاء ملف الوورد...")
doc = Document()
doc.styles["Normal"].font.name = "Times New Roman"
doc.styles["Normal"].font.size = Pt(12)

for elem in elements:
    if elem['type'] == 'image':
        x1, y1, x2, y2 = elem['bbox']

        # التأكد من أن الإحداثيات ضمن حدود الصورة
        x1, y1 = max(0, x1), max(0, y1)
        x2, y2 = min(w, x2), min(h, y2)

        cropped = pil_img.crop((x1, y1, x2, y2))
        img_byte_arr = io.BytesIO()
        cropped.save(img_byte_arr, format='PNG')

        para = doc.add_paragraph()
        para.alignment = WD_ALIGN_PARAGRAPH.CENTER
        run = para.add_run()

        # حساب العرض (افتراض دقة الشاشة 150 DPI كمتوسط لقطات الشاشة)
        img_width = (x2 - x1) / 150.0
        if img_width > 0.5:
            run.add_picture(io.BytesIO(img_byte_arr.getvalue()), width=Inches(img_width))

    elif elem['type'] == 'text':
        # التأكد من أن النص ليس داخل صورة تم اكتشافها
        is_inside_image = False
        for img_reg in image_regions:
            x_overlap = max(0, min(elem['bbox'][2], img_reg[2]) - max(elem['bbox'][0], img_reg[0]))
            y_overlap = max(0, min(elem['bbox'][3], img_reg[3]) - max(elem['bbox'][1], img_reg[1]))
            overlap_area = x_overlap * y_overlap
            elem_area = (elem['bbox'][2] - elem['bbox'][0]) * (elem['bbox'][3] - elem['bbox'][1])
            if elem_area > 0 and (overlap_area / elem_area) > 0.5:
                is_inside_image = True
                break

        if not is_inside_image:
            text = elem['content']
            para = doc.add_paragraph(text)
            para.alignment = WD_ALIGN_PARAGRAPH.LEFT

output_path = "/content/final_extracted_all.docx"
doc.save(output_path)
print(f"\n✅ تم الحفظ بنجاح في: {output_path}")
print("🎉 حمّل الملف وتأكد من وجود الرسومات والنصوص معاً.")
````
